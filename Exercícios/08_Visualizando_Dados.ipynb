###Questão 2

Apple (`AAPL`), Amazon (`AMZN`), Google (`GOOG`), Microsoft (`MSFT`) e Tesla (`TSLA`) são empresas gigantes da área de tecnologia que movimentam milhões de dólares na bolsa de valores americana.

Utilizando o pacote `yfinance`, faça:

> a) Obtenha os valores destas ações entre 01/01/2023 e 01/04/2023.
>
> b) Crie um gráfico de linha para estas ações.
>
> c) Crie um gráfico de dispersão para estas ações.

```python
# Questao 2
# Letra a)
# Carregando pacote
import yfinance as yf
# Definindo abreviação (ticker)
ticker = ["AAPL",  # Apple
          "AMZN",  # Amazon
          "GOOG",  # Google
          "MSFT",  # Microsoft
          "TSLA"]  # Tesla

# Obtendo dados
df_acoesEUA = yf.download(ticker,                # Código das ações
                          start="2023-01-01",    # Data inicial AAAA-MM-DD
                          end="2023-04-01")      # Data final AAAA-MM-DD

#Organizando o DF
df_acoesEUA = df_acoesEUA.iloc[:,[0,1,2,3,4]].round(2)
df_acoesEUA.columns = ['Apple', 'Amazon','Google','Microsoft','Tesla']

# Chamando DF
df_acoesEUA
```

```python
# Questao 2
# Letra b)

# Criando o gráfico
df_acoesEUA.plot()

# Adicionando:
plt.title('Valores de fechamento') # Título
plt.xlabel('Data')                 # Rótulo do eixo horizontal (x)
plt.ylabel('Valor (R$)')           # Rótulo do eixo vertical (y)

# Exibindo o gráfico
plt.show()
```

```python
# Questao 2
# Letra c)

# Criando o gráfico
pd.plotting.scatter_matrix(df_acoesEUA)

# Adicionando:
plt.xlabel('Data')                 # Rótulo do eixo horizontal (x)
plt.ylabel('Valor (R$)')           # Rótulo do eixo vertical (y)

# Exibindo o gráfico
plt.show()
```

###Questão 3

> A relação entre a capitalização de mercado das empresas listadas em uma bolsa de valores e o Produto Interno Bruto (PIB) de um país pode ser vista como um indicador econômico importante, e essa relação é frequentemente usada para análise e comparação.

Utilizando o pacote `pandas_datareader`, faça:

> a) Obtenha os dados sobre a *capitalização do mercado de ações em relação ao PIB dos EUA* (`DDDM01USA156NWDB`) e *China* (`DDDM01CNA156NWDB`) entre os anos de 2015 e 2020.
>
> b) Organize o `DataFrame`, ajuste o índice para exibir somente o ano e remova eventuais valores faltantes.
>
> c) Crie um gráfico de barras para comparação entre estes indicadores.

Links: https://fred.stlouisfed.org/series/DDDM01USA156NWDB

Links: https://fred.stlouisfed.org/series/DDDM01CNA156NWDB

```python
import pandas_datareader.data as pdr
import matplotlib.pyplot as plt

# Definindo código
codigo_fred = ['DDDM01CNA156NWDB','DDDM01USA156NWDB']

# Obtendo dados
df_fred = pdr.get_data_fred(
  codigo_fred,   # Lista de indicadores
  start='2015', # Data início AAAA
  end='2020')   # Data fim AAAA

# Resultado (tabela)
df_fred
```

```python
# Questao 3
# Letra b)
# Organize o DataFrame, ajuste o índice para exibir somente o ano e remova valores faltantes
df_fred = df_fred.dropna()  # Remova valores faltantes

# Renomeie o índice para os anos desejados
df_fred.index = ['2015', '2016', '2017', '2018', '2019']

# Escrevendo o nome das colunas
df_fred.columns = ['China', 'EUA']

df_fred
```

```python
# Questao 3
# Letra c)
# Criando o gráfico

# Crie um gráfico de barras para comparar os indicadores
df_fred.plot(kind='bar')

plt.title('Comparação de Indicadores')
plt.xlabel('Ano')
plt.ylabel('Valor')
plt.legend(df_fred.columns)

plt.show()
```
