###Questão 1

Apple (`AAPL`), Amazon (`AMZN`), Google (`GOOG`), Microsoft (`MSFT`) e Tesla (`TSLA`) são empresas gigantes da área de tecnologia que movimentam milhões de dólares na bolsa de valores americana.

A partir destes dados:

> a) Crie um resumo com os valores médios destas ações neste intervalo.
>
> b) Compute a volatilidade destas ações neste intervalo.
>
> c) Compute a matriz de correlação destas ações.

```python
# Questao 1
# A partir do DataFrame 'df_acoes', responda os demais itens:

# Carregando pacotes
import pandas as pd
import yfinance as yf

# Definindo abreviação (ticker)
tickers = ["AAPL",   # Apple
           "AMZN",   # Amazon,
           "GOOG",   # Google
           "MSFT",   # Microsoft Corporation
           "TSLA"]   # Tesla

# Obtendo dados
df_acoes = yf.download(tickers,       # Ações
                       start="2023-01-01",   # Data inicial AAAA-MM-DD
                       end="2023-04-01")     # Data final AAAA-MM-DD

# Selecionando as colunas `Adj Close` pela posição
df_acoes = df_acoes.iloc[:, [0, 1, 2, 3, 4]]

# Renomeando colunas
df_acoes.columns = ["AAPL", "AMZN", "GOOG", "MSFT", "TSLA"]

# Visualizando resultado
df_acoes
```

```python
# Questao 1
# Letra a)
medias = df_acoes.mean()
medias




```

```python
# Questao 1
# Letra b)
volatilidade = df_acoes.std()
volatilidade




```

```python
# Questao 1
# Letra c)
df_acoes.corr()




```

###Questão 2

> *Melhorar o acesso à eletricidade em comunidades carentes ou em regiões subdesenvolvidas é um passo importante para melhorar a qualidade de vida e aumentar a expectativa de vida das pessoas. É um lembrete de que o desenvolvimento humano é um processo complexo e interconectado, no qual diversos fatores, incluindo o acesso a serviços essenciais, desempenham um papel crucial.*

*   Expectativa de vida ao nascer - Brasil: https://data.worldbank.org/indicator/SP.DYN.LE00.IN?locations=BR
*   Acesso à eletricidade - Brasil: https://data.worldbank.org/indicator/EG.ELC.ACCS.ZS?locations=BR

A partir destes dados, responda:

> a) Qual o valor do coeficiente de correlação linear destes indicadores?
>
> b) Exiba esta relação de forma gráfica




```python
# Questao 2
# A partir do DataFrame 'df_wb', responda os demais itens:

# Carregando módulo (parte do pacote)
from pandas_datareader import wb

# Carregando pacote
import matplotlib.pyplot as plt

# Definindo código
codigo_wb = ['SP.DYN.LE00.IN',   # Expectativa de vida ao nascer (anos)
             'EG.ELC.ACCS.ZS']   # Acesso à eletricidade (% da população)

# Obtendo dados
df_wb = wb.download(indicator = codigo_wb,  # Lista de indicadores
                    country = ['BR'],       # Lista de países
                    start = 2011,           # Data início AAAA
                    end = 2020)             # Data fim AAAA

# Resultado (tabela)
df_wb
```

```python
# Questao 2
# Letra a)
df_wb.corr()




```

```python
# Questao 2
# Letra b)
# Criando o gráfico
pd.plotting.scatter_matrix(df_wb)

# Exibindo o gráfico
plt.show()




```
