###Questão 2

> O índice de Gini é fundamental para avaliar desigualdades econômicas ao quantificar a distribuição de renda ou riqueza na sociedade. Essa métrica é vital para governos e economistas, fornecendo insights sobre equidade, pobreza e estabilidade social, permitindo ações que promovam uma distribuição mais justa de recursos e desenvolvimento sustentável.

Utilizando o pacote `pandas_datareader`, obtenha toda a série histórica (ínício em 1981) deste índice para o Brasil. O código FRED para esta série é `SIPOVGINIBRA`.

> a. Renomeie a coluna para `Gini`
>
> b. Renomeie os índices somente com os anos
>
> c. Remova os dados faltantes
>
> d. Exiba o resultado na tela

Link: https://fred.stlouisfed.org/series/SIPOVGINIBRA

```python
# Questao 2

import pandas_datareader.data as pdr
import datetime

# Definindo código
codigo = ['SIPOVGINIBRA'] #Índice de Gini

# Definindo data: datetime.datetime(AAAA, MM, DD)
inicio = datetime.datetime(1981, 1, 1)
fim = datetime.datetime(2023, 1, 1)

# Obtendo dados
df_gini = pdr.DataReader(codigo, 'fred', inicio, fim)

# "Limpando" os dados
df_gini = df_gini.reset_index() #mantendo as series organizadas
df_gini.columns = ['Data','Gini'] #renomeando colunas
df_gini['Data'] = df_gini['Data'].dt.year #ajustando para somente o ano
df_gini = df_gini.rename(columns={'Data': 'Ano'}) #renomeando coluna de data
df_gini = df_gini.dropna() #removendo vazios


df_gini
```

###Questão 3

> O Produto Interno Bruto (PIB) per capita é uma métrica econômica que calcula a média da produção econômica de um país por habitante. Essa medida fornece uma visão importante do nível de prosperidade e bem-estar médio da população de um país, uma vez que considera a produção total em relação ao tamanho da população. O PIB per capita é amplamente usado para comparar o padrão de vida entre diferentes nações e ao longo do tempo, ajudando a identificar discrepâncias econômicas e a orientar políticas que visam melhorar o nível de vida das pessoas.

Utilizando o pacote `pandas_datareader`, obtenha a série histórica, a partir do ano 2000, desta métrica para o Brasil. O código WB para esta série é `NY.GDP.PCAP.CD`.

> a. Renomeie a coluna para `PIB`
>
> b. Reinicie os índices de forma que `country` e `year` passe a ser uma coluna do `DataFrame`
>
> c. Remova a coluna `country`
>
> d. Crie um novo `DataFrame` chamado `pib_inter`. Nele, insira somente os anos em que o valor do PIB foi superior à `8000` e inferior à `11000`.
>
> e. Exiba `pib_inter` na tela

Link: https://data.worldbank.org/indicator/NY.GDP.PCAP.CD?locations=BR

```python
# Questao 3

from pandas_datareader import wb

# Definindo código
codigo_wb = ['NY.GDP.PCAP.CD']   #Produto Interno Bruto (PIB)

# Obtendo dados
df_wb = wb.download(indicator = codigo_wb,  # Lista de indicadores
country = ['BR'],       # Lista de países
start = 2000,           # Data início AAAA
end = 2023)             # Data fim AAAA

#tratamento do conteúdo:
df_wb = df_wb.reset_index() #Reinicie os índices de forma que country e year passe a ser uma coluna do DataFrame
df_wb = df_wb.iloc[:,[1,2]] #Remova a coluna country
df_wb.columns = ['Ano','PIB'] #Renomeie a coluna para PIB

pib_inter = df_wb.copy() #Crie um novo DataFrame chamado pib_inter. Prefiri usar a função copy para não perder o dado anterior
pib_inter = pib_inter[(pib_inter['PIB'] > 8000) & (pib_inter['PIB'] < 11000)] #anos em que o valor do PIB foi superior à 8000 e inferior à 11000.

#Exiba pib_inter na tela
pib_inter


```
