**(1,0 ponto)** 1.b) Todos os novos clientes devem ser classificados conforme o seu **tipo de perfil de investidor**:
* Também chamado de _suitability_, o perfil de investidor é uma espécie de análise que identifica suas preferências e expectativas em relação aos riscos nos investimentos.

> Crie uma função da qual:
> *  O argumento de entrada seja `perfil`
>   *  Se `perfil = conservador`, retorne o seguinte texto:
>     * _Sugestões de investimento para perfil `perfil`: Certificado de Depósito Bancário (CDB), Letras Financeiras (LF) e Tesouro Direto (TD)_
>   *  Se `perfil = moderado`, retorne o seguinte texto:
>     * _Sugestões de investimento para perfil `perfil`: Letras de Crédito Imobiliário (LCI), Letras de Crédito do Agronegócio (LCA) e Debêntures_
>   *  Se `perfil = arrojado`, retorne o seguinte texto:
>     * _Sugestões de investimento para perfil `perfil`: Fundos Multimercado, Ações e Exchange Traded Fund (ETF)_

Teste a função utilizando um perfil de preferência.

```python
# Questao 1
# 1.b)

# Funcao
def perfil(escolha):
    #retorna a melhor escolha de investimento de acordo com o perfil de investidor, sendo 1 para conservador, 2 para moderado e 3 para arrojado.

    sugestão = "\033[93mNossa sugestão de investimento é:\033[0m "

    if escolha == 1:
        return sugestão + "Certificado de Depósito Bancário (CDB), Letras Financeiras (LF) e Tesouro Direto (TD)"
    elif escolha == 2:
        return sugestão + "Letras de Crédito Imobiliário (LCI), Letras de Crédito do Agronegócio (LCA) e Debêntures"
    elif escolha == 3:
        return sugestão + "Fundos Multimercado, Ações e Exchange Traded Fund (ETF)"
    else:
        return "\033[31mSua opção não é válida\033[m"

# Teste
escolha = int(input('''Escolha o perfil de investidor:

[ 1 ] Conservador
[ 2 ] Moderado
[ 3 ] Arrojado

'''))

print(f"\n {perfil(escolha)}")
```

**(1,0 ponto)** 1.c) É importante acompanhar em tempo real a evolução dos retornos dos investimentos, principalmente no caso de ações.
*  **Retorno percentual** é a diferença entre o `valor atual` e o `valor pago`: $\frac{atual - pago}{pago} \times 100\%$
> Crie uma função da qual:
> *  Os argumentos de entrada sejam `pago` e `atual`
> * Retorne o **retorno percentual** do investimento

Teste a função sabendo que o cliente investiu 153,67 reais e no momento, o valor de mercado deste investimento é 167,28 reais.

```python
# Funcao
def retorno(p, a):
    # Retorna o valor percentual do investimento
    r = (((a - p) / p) * 100)
    if r>=0:
      return "\033[92m{:.2f}%\033[m".format(r)
    if r<=0:
      return "\033[31m{:.2f}%\033[m".format(r)

# Teste
p = 153.67
a = 167.28

print(f"O retorno do seu investimento é de {retorno(p, a)}")
```

> _O Mercosul, ou Mercado Comum do Sul, é um bloco econômico e político sul-americano criado em 1991. Seu objetivo é criar um mercado comum entre os países membros, com a livre circulação de bens, serviços, pessoas, capitais e informações. O bloco também busca promover a cooperação em áreas como educação, cultura, ciência e tecnologia._
>
> _Saiba mais: https://www.mercosur.int/pt-br/quem-somos/paises-do-mercosul/_
>
> Os países participantes são:
> * Países parte
>   * Argentina (`AR`)
>   * Brasil (`BR`)
>   * Paraguai (`PY`)
>   * Uruguai (`UY`)
>   * Venezuela* (`VE`)
> * Países associados
>   * Bolívia (`BO`)
>   * Chile (`CL`)
>   * Colômbia (`CO`)
>   * Equador (`EC`)
>   * Guiana (`GY`)
>   * Peru (`PE`)
>   * Suriname (`SR`)
>
> _Um importante indicador para o planejamento econômico e social destes países é a **expectativa de vida ao nascer** (`SP.DYN.LE00.IN`). Um país com uma população com expectativa de vida alta terá uma população mais ativa e produtiva, o que pode contribuir para o crescimento econômico e atrair melhores investimentos externos._

**(1,0 ponto)** 4.a) Utilizando o módulo `wb` do pacote `pandas_datareader`, obtenha a _expectativa de vida ao nascer_ (`SP.DYN.LE00.IN`) de todos os países do Mercosul para o ano de 2021.

*   Reinicie os índices das linhas
*   Atribua a variável `country` como o novo índice das linhas
*   Utilizando o método `.loc`, selecione somente a coluna `SP.DYN.LE00.IN`

```python
# Carregando pacote `Pandas`
import pandas as pd
# Carregando módulo (parte do pacote)
from pandas_datareader import wb

# Definindo código
codigo_wb = ['SP.DYN.LE00.IN']   # Expectativa de vida ao nascer (anos)
paises = ['AR', 'BR', 'PY', 'UY', 'VE', 'BO', 'CL', 'CO', 'EC', 'GY', 'PE', 'SR']

# Obtendo dados
tb_expectativa = wb.download(indicator=codigo_wb,  # Lista de indicadores
                             country=paises,       # Lista de países
                             start=2021,           # Data início AAAA
                             end=2021)             # Data fim AAAA

# Reinicie os índices das linhas
tb_expectativa = tb_expectativa.reset_index()

# Atribua a variável country como o novo índice das linhas
tb_expectativa = tb_expectativa.set_index('country')

# Utilizando o método .loc, selecione somente a coluna SP.DYN.LE00.IN
tb_expectativa = tb_expectativa.loc[:, 'SP.DYN.LE00.IN']

# Resultado (tabela)
tb_expectativa


```

**(1,0 ponto)** 4.b) Construa um gráfico de barras do indicador em questão para estes países.

*   Adicione adequadamente título e rótulos nos eixos

```python
# Criando o gráfico
#deixando em ordem para ficar mais organizado e visualmente mais fácil de ler:
tb_expectativa = tb_expectativa.sort_values(ascending=False)

#plotando o gráfico
tb_expectativa.plot.bar(rot=45,              # Angulo da rotação dos rótulos na horizontal > 0 = sem rotação
                 edgecolor='black',          # Cor das bordas
                 color = 'gold',             # Cor do Gráfico
                 grid = False)               # Remove as linhas de grade

# Adicionando:
plt.title('Expectativa de Vida em 2021 - Mercosul')      # Título
plt.xlabel('País')                  # Rótulo do eixo horizontal (x)
plt.ylabel('Anos')       # Rótulo do eixo vertical (y)

# Exibindo o gráfico
plt.show()
```
