## Agrupamento nas AnÃ¡lises do Mercado Financeiro

ðŸ—‚ï¸ O agrupamento (_clustering_) no mercado financeiro Ã© essencial para uma sÃ©rie de aplicaÃ§Ãµes, desde a segmentaÃ§Ã£o de clientes e a construÃ§Ã£o de carteiras de investimento atÃ© a detecÃ§Ã£o de anomalias e a previsÃ£o de tendÃªncias de mercado. > Em resumo, o agrupamento desempenha um papel fundamental na anÃ¡lise financeira, ajudando a extrair informaÃ§Ãµes valiosas dos dados, identificar padrÃµes, reduzir riscos e tomar decisÃµes informadas no mercado financeiro. ## CaracterÃ­sticas do agrupamento

ðŸ“Œ _Clustering_, em portuguÃªs _"agrupamento"_ ou _"clusterizaÃ§Ã£o"_, Ã© uma tÃ©cnica de anÃ¡lise de dados e aprendizado de mÃ¡quina nÃ£o supervisionado que envolve a organizaÃ§Ã£o de um conjunto de dados em grupos ou _"clusters"_ com base em caracterÃ­sticas semelhantes. O objetivo principal do _clustering_ Ã© dividir os dados em grupos de modo que os itens dentro de cada grupo sejam mais semelhantes entre si do que com os itens em outros grupos.

Aqui estÃ£o algumas caracterÃ­sticas-chave do _clustering_:

*   **NÃ£o supervisionado**: O _clustering_ Ã© um mÃ©todo nÃ£o supervisionado, o que significa que nÃ£o requer rÃ³tulos ou categorias predefinidas nos dados. Em vez disso, ele identifica automaticamente padrÃµes e estruturas nos dados com base nas caracterÃ­sticas intrÃ­nsecas.

*   **Similaridade**: A similaridade entre os pontos de dados Ã© a base para a formaÃ§Ã£o de clusters. Os algoritmos de _clustering_ calculam a similaridade entre os pontos de dados.

*    **Objetivo**: O objetivo principal do _clustering_ Ã© agrupar dados de forma que os itens dentro de um _cluster_ sejam mais parecidos entre si do que com os itens em outros _clusters_.

*    **AplicaÃ§Ãµes**: O _clustering_ Ã© amplamente utilizado em vÃ¡rias Ã¡reas, incluindo anÃ¡lise de dados, mineraÃ§Ã£o de dados, segmentaÃ§Ã£o de clientes, agrupamento de documentos, reconhecimento de padrÃµes, biologia, anÃ¡lise de imagem e muito mais.

```python
# Exemplo introdutÃ³rio: Entendendo como o agrupamento Ã© feito.

!pip install descartes
!pip install geobr

# Carregando pacotes
import geopandas as gpd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import geobr

# Dados sobre as capitais do Brasil
data = [
["Norte", "Rondonia", "RO", -63.58061, -11.5057341],
["Norte", "Acre", "AC", -70.81200, -9.0237964],
["Norte", "Amazonas", "AM", -65.85606, -3.4168427],
["Norte", "Roraima", "RR", -62.07510, 2.7375971],
["Norte", "Para", "PA", -54.93062, -1.9981271],
["Norte", "Amapa", "AP", -52.00296, 0.9019925],
["Norte", "Tocantins", "TO", -48.29825, -10.1752800],
["Nordeste", "Maranhao", "MA", -45.27442, -4.9609498],
["Nordeste", "Piaui", "PI", -42.72892, -7.7183401],
["Nordeste", "Ceara", "CE", -39.32062, -5.4983977],
["Nordeste", "Rio Grande do Norte", "RN", -36.95411, -5.4025803],
["Nordeste", "Paraiba", "PB", -36.78195, -7.2399609],
["Nordeste", "Pernambuco", "PE", -36.95411, -8.8137173],
["Nordeste", "Alagoas", "AL", -36.78195, -9.5713058],
["Nordeste", "Sergipe", "SE", -37.38566, -10.5740934],
["Nordeste", "Bahia", "BA", -41.70073, -12.5797380],
["Sudeste", "Minas Gerais", "MG", -44.55503, -18.5121780],
["Sudeste", "Espirito Santo", "ES", -40.30886, -19.1834229],
["Sudeste", "Rio de Janeiro", "RJ", -43.17290, -22.9068467],
["Sudeste", "Sao Paulo", "SP", -46.63331, -23.5505199],
["Sul", "Parana", "PR", -52.02154, -25.2520888],
["Sul", "Santa Catarina", "SC", -50.21886, -27.2423392],
["Sul", "Rio Grande do Sul", "RS", -51.21770, -30.0346316],
["Centro-Oeste", "Mato Grosso do Sul", "MS", -54.78515, -20.7722295],
["Centro-Oeste", "Mato Grosso", "MT", -56.92110, -12.6818712],
["Centro-Oeste", "Goias", "GO", -49.83622, -15.8270369],
["Centro-Oeste", "Distrito Federal", "DF", -47.86447, -15.7997654]
]

# Nomes das colunas
columns = ["region", "state", "abbrev_state", "long", "lat"]

# Montando mapa do Brasil
brasil = (geobr.
          read_state(year=2019).
          merge(pd.DataFrame(data, columns=columns),
                on = 'abbrev_state'))

# Criando figura
fig, ax = plt.subplots()

# Lista de cores
region_colors = ['tab:blue', 'tab:orange', 'tab:green',
                 'tab:red', 'tab:purple']

# Desenhando cada regiÃ£o
regions = brasil['region'].unique()

for i, region in enumerate(regions):
    region_data = brasil.loc[brasil['region'] == region, :]
    region_data.plot(ax=ax, edgecolor='black',
                     facecolor=region_colors[i],
                     linewidth=0.5,
                     label=region)
    plt.scatter(region_data['long'],
                region_data['lat'],
                label=region,
                edgecolors='black')

# Adicionando rÃ³tulos
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title('RegiÃµes do Brasil')
ax.legend()

# Show the plot
plt.show()
```

ðŸ‘‰ Essas sÃ£o as regiÃµes do Brasil.

* Os cÃ­rculos representam a localizaÃ§Ã£o das capitais de cada Estado.

ðŸ‘‰ Para criar o agrupamento, utilizaremos as ferramentas do pacote `sklearn`

```python
# Carregando pacote e partes necessarias
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
import numpy as np

# Criando um DataFrame com as informaÃ§Ãµes que serÃ£o utilizadas como mÃ©tricas
df_regioes = brasil[['long',	'lat']]
df_regioes.index = brasil['name_state']
df_regioes
```

> âš ï¸ SÃ£o permitidas somente variÃ¡veis numÃ©ricas para que sejam realizados os cÃ¡lculos.

ðŸ‘‰ Em `df_regioes`, estÃ£o contidas a `longitude` e `latitude` das capitais (26 Estados e Distrito Federal)

```python
# Interligando passos
pipeline = make_pipeline(
    StandardScaler(),          # Padronizando dados
    KMeans(n_clusters=5,       # Num de grupos
           random_state = 321, # Semente aleatÃ³ria
           n_init = "auto")    # Mantenha essa opÃ§Ã£o no "auto"
)

# Agrupamento
grupos = pipeline.fit(df_regioes) # DataFrame com as mÃ©tricas

# Exibindo resultado
print(grupos.predict(df_regioes))
```

> âš ï¸ **ANALISANDO CADA ITEM**

`n_clusters` = NÃºmero de grupos (*cluster* = grupo em portuguÃªs)

`random_state` = Ã‰ um nÃºmero qualquer que utilizamos para que o algoritmo se inicie sempre da mesma forma. Pode ser o valor que vocÃª preferir, aqui utilizamos `123`. Tente outro valor. Possivelmente, o resultado serÃ¡ diferente do obtido.

`grupos.predict()` = Ã‰ a previsÃ£o/rÃ³tulo de cada grupo predito para cada observaÃ§Ã£o.

> âš ï¸ **INTERPRETANDO**

Este mÃ©todo simplesmente aloca as informaÃ§Ãµes na quantidade de grupos prÃ© determinada em `n_clusters`.

ðŸ‘‰ Cabe a nÃ³s interpretar o que significa cada grupo.

```python
# Juntando a regiao calculada com a localizaÃ§Ã£o
df_regioes['regiao'] = grupos.predict(df_regioes)
df_regioes
```

ðŸ‘‰ Com a classificaÃ§Ã£o criada, podemos visualizar como este mÃ©todo agruparia os Estados do Brasil em 5 regiÃµes:

```python
# Criando figura
fig, ax = plt.subplots()

# Lista de cores
region_colors = ['tab:blue', 'tab:orange', 'tab:green',
                 'tab:red', 'tab:purple']

# Desenhando cada regiÃ£o
regions = brasil['region'].unique()

for i, region in enumerate(regions):
    region_data = brasil.loc[brasil['region'] == region, :]
    region_data.plot(ax=ax, edgecolor='black',
                     facecolor=region_colors[i],
                     linewidth=0.5,
                     label=region)

for i, region in enumerate(regions):
    class_data = df_regioes.loc[df_regioes['regiao'] == i, :]
    plt.scatter(class_data['long'],
                class_data['lat'],
                marker=f'${i}$',
                edgecolors='black')

# Adicionando rÃ³tulos
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title('Agrupamento automÃ¡tico das RegiÃµes do Brasil')
ax.legend()

# Show the plot
plt.show()
```

> âš ï¸ **ATENÃ‡ÃƒO**

* Os nomes do grupos (neste caso, representado por nÃºmeros de 0 Ã  4) sÃ£o permutÃ¡veis. Ou seja, o grupo `0` pode ser chamado de `1` e vice-versa toda vez que alterarmos o argumento `random_state`.

* NÃ£o leve tÃ£o a sÃ©rio este exemplo! Estamos exclusivamente considerando a localizaÃ§Ã£o das capitais. Sabemos que as regiÃµes foram formadas considerando outros aspectos polÃ­ticos e econÃ´micos.

> ðŸ› ï¸ *Exemplo prÃ¡tico 1:* Agrupando e classificando aÃ§Ãµes

> _O *retorno percentual* pode ser utilizado como um indicador de valorizaÃ§Ã£o de uma aÃ§Ã£o. Por outro lado, o *volume* pode ser visto como um indicador de liquidez (ou seja, Ã© mais fÃ¡cil comprar e vender aÃ§Ãµes sem afetar significativamente o preÃ§o de mercado)_

ðŸ‘‰ Utilizando estes 2 indicadores, vamos agrupar 23 aÃ§Ãµes em 2 grupos considerando o 1o trimestre de 2023

```python
# Carregando pacote
import yfinance as yf

# Tickers
tickers = ["ABEV3.SA", "AMER3.SA", "ASAI3.SA", "B3SA3.SA", "BBDC4.SA",
           "BRFS3.SA", "CIEL3.SA", "COGN3.SA", "CPLE6.SA", "CSNA3.SA",
           "CVCB3.SA", "HAPV3.SA", "IFCM3.SA", "ITSA4.SA", "ITUB4.SA",
           "MGLU3.SA", "PETR4.SA", "RAIL3.SA", "SEQL3.SA", "SOMA3.SA",
           "USIM5.SA", "VALE3.SA", "VIIA3.SA"]

# Obtendo dados
df_retorno = yf.download(tickers,                # AÃ§Ã£o
                         start="2023-01-01",     # Data inicial AAAA-MM-DD
                         end="2023-04-01",       # Data final AAAA-MM-DD
                         progress=False).iloc[:, list(range(23))]
                         # Selecionando as 23 primeiras colunas

df_volume = yf.download(tickers,                # AÃ§Ã£o
                        start="2023-01-01",     # Data inicial AAAA-MM-DD
                        end="2023-04-01",       # Data final AAAA-MM-DD
                        progress=False).iloc[:, list(range(-23, 0))]
                        # Selecionando as 23 ultimas colunas

# Renomeando colunas
df_retorno.columns = tickers
df_volume.columns = tickers
```

```python
# Visualizando resultado
df_retorno
```

```python
# Visualizando resultado
df_volume
```

ðŸ‘‰ Precisamos calcular o `retorno percentual = valor atual / valor anterior * 100`. Utilzaremos o mÃ©todo `.pct_change()`

```python
# Calculando retorno percentual
# > Para cada colunas de `df_retorno`,
# > substituo o valor pelo retorno percentual
for col in df_retorno.columns:
  df_retorno[col] = df_retorno[col].pct_change() * 100

# Visualizando resultado
df_retorno
```

ðŸ‘‰ A primeira linha Ã© inviabilizada por conta deste cÃ¡lculo. Vamos removÃª-la nos 2 `DataFrames`

```python
# Removendo a primeira linha em ambos
df_retorno = df_retorno.iloc[1:]
df_volume = df_volume.iloc[1:]
```

ðŸ‘‰ Precisamos de um Ãºnico valor para representar todo o trimestre. Vamos utilizar a **mÃ©dia** para resumir toda a informaÃ§Ã£o

```python
# Crie um novo DataFrame com a mÃ©dia das colunas
df_indicadores = pd.DataFrame([df_retorno.mean(),
                               df_volume.mean() / 100000], # DivisÃ£o por 100.000
                                                           # para melhor visualizaÃ§Ã£o
                              index = ['Retorno', 'Volume'],
                              columns=tickers)

# Invertendo linhas por colunas
df_indicadores = df_indicadores.transpose()

# Visualizando resultado
df_indicadores
```

ðŸ‘‰ Da mesma forma que fizemos no exemplo introdutÃ³rio, vamos compreender estes 2 indicadores como 'localizaÃ§Ãµes' em um mapa

```python
# Carregando pacotes
import matplotlib.pyplot as plt

# GrÃ¡fico de dispersÃ£o
plt.scatter(df_indicadores['Retorno'], df_indicadores['Volume'], c='black')

# Adicionando rÃ³tulos aos pontos
for i, txt in enumerate(tickers):
    plt.annotate(txt, (df_indicadores['Retorno'][i], df_indicadores['Volume'][i]),
                 fontsize=8, ha='left', c='grey')

# Definindo rÃ³tulos dos eixos
plt.xlabel('Retorno')
plt.ylabel('Volume')

# Definindo o tÃ­tulo do grÃ¡fico
plt.title('Retorno vs. Volume')

# Exibindo o grÃ¡fico
plt.grid(True)
plt.show()
```

ðŸ‘‰ Finalmente, vamos criar 2 grupos para estas aÃ§Ãµes

```python
# Carregando pacote e partes necessarias
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
import numpy as np

# Interligando passos
pipeline = make_pipeline(
    StandardScaler(),          # Padronizando dados
    KMeans(n_clusters=2,       # Num de grupos
           random_state = 999, # Semente aleatÃ³ria (pode usar a que vocÃª quiser)
           n_init = "auto")    # Mantenha essa opÃ§Ã£o no "auto"
)

# Agrupamento
grupos_acoes = pipeline.fit(df_indicadores) # DataFrame com as mÃ©tricas

# Juntando o grupo calculado com os indicadores
df_indicadores['Grupo'] = grupos_acoes.predict(df_indicadores)
df_indicadores
```

ðŸ‘‰ Vamos visualizar o resultado

```python
# GrÃ¡fico de dispersÃ£o
plt.scatter(df_indicadores['Retorno'], df_indicadores['Volume'], c='white')

# Lista de cores
cores_grupo = ['tab:blue', 'tab:orange']

# Adicionando rÃ³tulos aos pontos
for i, txt in enumerate(tickers):
    plt.annotate(txt, (df_indicadores['Retorno'][i], df_indicadores['Volume'][i]),
                 fontsize=8, ha='left', c='grey')
    plt.scatter(df_indicadores['Retorno'][i],
                df_indicadores['Volume'][i],
                c=cores_grupo[df_indicadores['Grupo'][i]])

# Definindo rÃ³tulos dos eixos
plt.xlabel('Retorno')
plt.ylabel('Volume')

# Definindo o tÃ­tulo do grÃ¡fico
plt.title('Retorno vs. Volume')

# Exibindo o grÃ¡fico
plt.grid(True)
plt.show()
```

ðŸ‘‰ Vamos replicar agora para 4 grupos e visualizar o resultado

```python
# Interligando passos
pipeline = make_pipeline(
    StandardScaler(),          # Padronizando dados
    KMeans(n_clusters=4,       # Num de grupos
           random_state = 999, # Semente aleatÃ³ria (pode usar a que vocÃª quiser)
           n_init = "auto")    # Mantenha essa opÃ§Ã£o no "auto"
)

# Agrupamento
grupos_acoes = pipeline.fit(df_indicadores) # DataFrame com as mÃ©tricas

# Juntando o grupo calculado com os indicadores
df_indicadores['Grupo'] = grupos_acoes.predict(df_indicadores)
df_indicadores

# GrÃ¡fico de dispersÃ£o
plt.scatter(df_indicadores['Retorno'], df_indicadores['Volume'], c='white')

# Lista de cores
cores_grupo = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red']

# Adicionando rÃ³tulos aos pontos
for i, txt in enumerate(tickers):
    plt.annotate(txt, (df_indicadores['Retorno'][i], df_indicadores['Volume'][i]),
                 fontsize=8, ha='left', c='grey')
    plt.scatter(df_indicadores['Retorno'][i],
                df_indicadores['Volume'][i],
                c=cores_grupo[df_indicadores['Grupo'][i]])

# Definindo rÃ³tulos dos eixos
plt.xlabel('Retorno')
plt.ylabel('Volume')

# Definindo o tÃ­tulo do grÃ¡fico
plt.title('Retorno vs. Volume')

# Exibindo o grÃ¡fico
plt.grid(True)
plt.show()
```

> ðŸ‘‰ Vamos utilizar estes indicadores (2021) para classificar paÃ­ses

```python
# Carregando mÃ³dulo (parte do pacote)
from pandas_datareader import wb

siglas_paises = [
    'CA',  # CanadÃ¡
    'MX',  # MÃ©xico
    'GB',  # Reino Unido
    'FR',  # FranÃ§a
    'DE',  # Alemanha
    'IT',  # ItÃ¡lia
    'ES',  # Espanha
    'PT',  # Portugal
    'AU',  # AustrÃ¡lia
    'NZ',  # Nova ZelÃ¢ndia
    'JP',  # JapÃ£o
    'CN',  # China
    'IN',  # Ãndia
    'BR',  # Brasil
    'AR',  # Argentina
    'CL',  # Chile
    'CO',  # ColÃ´mbia
    'PE',  # Peru
    'EG',  # Egito
    'ZA',  # Ãfrica do Sul
    'NG',  # NigÃ©ria
    'KE',  # QuÃªnia
    'RU',  # RÃºssia
    'TR',  # Turquia
    'SA',  # ArÃ¡bia Saudita
    'AE',  # Emirados Ãrabes Unidos
    'KR',  # Coreia do Sul
    'US'   # Estados Unidos
]

# Obtendo dados
df_wb = wb.download(indicator = ['NY.GDP.PCAP.CD',    # PIB per capita (US$)
                                 'SL.UEM.TOTL.ZS'],   # Desemprego (%)
                    country = siglas_paises,          # Lista de paÃ­ses
                    start = 2021,                     # Data inÃ­cio AAAA
                    end = 2021)                       # Data fim AAAA

# Editando DataFrame
df_wb = df_wb.reset_index()
df_wb.index = df_wb['country']
df_wb = df_wb.loc[:, ['NY.GDP.PCAP.CD', 'SL.UEM.TOTL.ZS']]
df_wb.columns = ['pib', 'desemprego']
df_wb['pib'] = df_wb['pib'] / 1000

# Resultado (tabela)
df_wb
```

ðŸ‘‰ Vamos visualizar como os paÃ­ses se distribuem a partir destas mÃ©tricas

```python
# GrÃ¡fico de dispersÃ£o
plt.scatter(df_wb['pib'], df_wb['desemprego'], c='black')

# Adicionando rÃ³tulos aos pontos
for i, txt in enumerate(df_wb.index):
    plt.annotate(txt, (df_wb['pib'][i], df_wb['desemprego'][i]),
                 fontsize=8, ha='left', c='grey')

# Definindo rÃ³tulos dos eixos
plt.xlabel('PIB per capita (US$)')
plt.ylabel('Desemprego (%)')

# Exibindo o grÃ¡fico
plt.grid(True)
plt.show()
```

```python
# Interligando passos
pipeline = make_pipeline(
    StandardScaler(),          # Padronizando dados
    KMeans(n_clusters=5,       # Num de grupos
           random_state = 777, # Semente aleatÃ³ria (pode usar a que vocÃª quiser)
           n_init = "auto")    # Mantenha essa opÃ§Ã£o no "auto"
)

# Agrupamento
grupos_paises = pipeline.fit(df_wb) # DataFrame com as mÃ©tricas

# Juntando o grupo calculado com os indicadores
df_wb['grupo'] = grupos_paises.predict(df_wb)
df_wb

# GrÃ¡fico de dispersÃ£o
plt.scatter(df_wb['pib'], df_wb['desemprego'], c='black')

# Lista de cores
cores_grupo = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red',
               'tab:purple']

# Adicionando rÃ³tulos aos pontos
for i, txt in enumerate(df_wb.index):
    plt.annotate(txt, (df_wb['pib'][i], df_wb['desemprego'][i]),
                 fontsize=8, ha='left', c='grey')
    plt.scatter(df_wb['pib'][i],
                df_wb['desemprego'][i],
                c=cores_grupo[df_wb['grupo'][i]])

# Definindo rÃ³tulos dos eixos
plt.xlabel('PIB per capita (US$)')
plt.ylabel('Desemprego (%)')

# Exibindo o grÃ¡fico
plt.grid(True)
plt.show()
```
