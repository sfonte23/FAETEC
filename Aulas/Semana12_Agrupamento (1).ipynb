## Agrupamento nas Análises do Mercado Financeiro

🗂️ O agrupamento (_clustering_) no mercado financeiro é essencial para uma série de aplicações, desde a segmentação de clientes e a construção de carteiras de investimento até a detecção de anomalias e a previsão de tendências de mercado. Neste contexto, o _clustering_ atua como uma lente que ajuda a revelar padrões ocultos nos dados financeiros, oferecendo insights que podem orientar estratégias de investimento, gerenciamento de riscos e tomada de decisões informadas. Algumas aplicações:

1. **Segmentação de Mercado**: O _clustering_ pode ajudar a identificar segmentos de mercado com características financeiras semelhantes. Isso é crucial para instituições financeiras, corretoras e investidores que desejam direcionar estratégias de investimento ou produtos específicos para grupos de clientes com necessidades financeiras semelhantes.

2. **Análise de Carteira**: Investidores e gestores de fundos usam o _clustering_ para agrupar ativos financeiros semelhantes, como ações, títulos ou moedas, para construir carteiras de investimento diversificadas e equilibradas. Isso ajuda a mitigar riscos e otimizar os retornos.

3. **Detecção de Anomalias**: O _clustering_ também é útil na detecção de anomalias financeiras. Por exemplo, pode ser usado para identificar transações suspeitas em instituições financeiras, ajudando a combater a lavagem de dinheiro e fraudes.

4. **Previsão de Mercado**: Algoritmos de _clustering_ podem ser usados para identificar padrões em séries temporais de preços de ativos financeiros. Isso pode ajudar os traders a tomar decisões informadas sobre a direção dos mercados.

5. **Gerenciamento de Riscos**: O _clustering_ pode ser aplicado para avaliar o risco em carteiras de investimento. Isso permite que os investidores identifiquem e mitiguem os riscos associados a determinados grupos de ativos.

6. **Análise de Crédito**: Instituições financeiras usam o _clustering_ para avaliar o risco de crédito, agrupando os clientes de acordo com seu histórico financeiro e perfil de crédito. Isso ajuda na tomada de decisões de empréstimos e na determinação de taxas de juros apropriadas.

7. **Identificação de Tendências**: O _clustering_ pode ser aplicado para identificar tendências emergentes nos mercados financeiros. Isso permite que os investidores se posicionem de forma proativa em relação a oportunidades ou riscos emergentes.

8. **Personalização de Serviços**: As instituições financeiras podem usar o _clustering_ para personalizar serviços financeiros, como oferecer produtos específicos com base nos padrões de gastos e investimentos dos clientes.

> Em resumo, o agrupamento desempenha um papel fundamental na análise financeira, ajudando a extrair informações valiosas dos dados, identificar padrões, reduzir riscos e tomar decisões informadas no mercado financeiro. Essa técnica permite uma compreensão mais profunda dos dados financeiros e pode ser uma ferramenta poderosa para instituições financeiras, investidores e gestores de ativos.

## Características do agrupamento

📌 _Clustering_, em português _"agrupamento"_ ou _"clusterização"_, é uma técnica de análise de dados e aprendizado de máquina não supervisionado que envolve a organização de um conjunto de dados em grupos ou _"clusters"_ com base em características semelhantes. O objetivo principal do _clustering_ é dividir os dados em grupos de modo que os itens dentro de cada grupo sejam mais semelhantes entre si do que com os itens em outros grupos.

Aqui estão algumas características-chave do _clustering_:

*   **Não supervisionado**: O _clustering_ é um método não supervisionado, o que significa que não requer rótulos ou categorias predefinidas nos dados. Em vez disso, ele identifica automaticamente padrões e estruturas nos dados com base nas características intrínsecas.

*   **Similaridade**: A similaridade entre os pontos de dados é a base para a formação de clusters. Os algoritmos de _clustering_ calculam a similaridade entre os pontos de dados.

*    **Objetivo**: O objetivo principal do _clustering_ é agrupar dados de forma que os itens dentro de um _cluster_ sejam mais parecidos entre si do que com os itens em outros _clusters_.

*    **Aplicações**: O _clustering_ é amplamente utilizado em várias áreas, incluindo análise de dados, mineração de dados, segmentação de clientes, agrupamento de documentos, reconhecimento de padrões, biologia, análise de imagem e muito mais.

👉 Imagine a tarefa de recriar essas mesmas 5 regiões considerando **somente** a localização (latitude e longitude) das capitais. Podemos utilizar _clustering_.

```python
# Exemplo introdutório: Entendendo como o agrupamento é feito.

!pip install descartes
!pip install geobr

# Carregando pacotes
import geopandas as gpd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import geobr

# Dados sobre as capitais do Brasil
data = [
["Norte", "Rondonia", "RO", -63.58061, -11.5057341],
["Norte", "Acre", "AC", -70.81200, -9.0237964],
["Norte", "Amazonas", "AM", -65.85606, -3.4168427],
["Norte", "Roraima", "RR", -62.07510, 2.7375971],
["Norte", "Para", "PA", -54.93062, -1.9981271],
["Norte", "Amapa", "AP", -52.00296, 0.9019925],
["Norte", "Tocantins", "TO", -48.29825, -10.1752800],
["Nordeste", "Maranhao", "MA", -45.27442, -4.9609498],
["Nordeste", "Piaui", "PI", -42.72892, -7.7183401],
["Nordeste", "Ceara", "CE", -39.32062, -5.4983977],
["Nordeste", "Rio Grande do Norte", "RN", -36.95411, -5.4025803],
["Nordeste", "Paraiba", "PB", -36.78195, -7.2399609],
["Nordeste", "Pernambuco", "PE", -36.95411, -8.8137173],
["Nordeste", "Alagoas", "AL", -36.78195, -9.5713058],
["Nordeste", "Sergipe", "SE", -37.38566, -10.5740934],
["Nordeste", "Bahia", "BA", -41.70073, -12.5797380],
["Sudeste", "Minas Gerais", "MG", -44.55503, -18.5121780],
["Sudeste", "Espirito Santo", "ES", -40.30886, -19.1834229],
["Sudeste", "Rio de Janeiro", "RJ", -43.17290, -22.9068467],
["Sudeste", "Sao Paulo", "SP", -46.63331, -23.5505199],
["Sul", "Parana", "PR", -52.02154, -25.2520888],
["Sul", "Santa Catarina", "SC", -50.21886, -27.2423392],
["Sul", "Rio Grande do Sul", "RS", -51.21770, -30.0346316],
["Centro-Oeste", "Mato Grosso do Sul", "MS", -54.78515, -20.7722295],
["Centro-Oeste", "Mato Grosso", "MT", -56.92110, -12.6818712],
["Centro-Oeste", "Goias", "GO", -49.83622, -15.8270369],
["Centro-Oeste", "Distrito Federal", "DF", -47.86447, -15.7997654]
]

# Nomes das colunas
columns = ["region", "state", "abbrev_state", "long", "lat"]

# Montando mapa do Brasil
brasil = (geobr.
          read_state(year=2019).
          merge(pd.DataFrame(data, columns=columns),
                on = 'abbrev_state'))

# Criando figura
fig, ax = plt.subplots()

# Lista de cores
region_colors = ['tab:blue', 'tab:orange', 'tab:green',
                 'tab:red', 'tab:purple']

# Desenhando cada região
regions = brasil['region'].unique()

for i, region in enumerate(regions):
    region_data = brasil.loc[brasil['region'] == region, :]
    region_data.plot(ax=ax, edgecolor='black',
                     facecolor=region_colors[i],
                     linewidth=0.5,
                     label=region)
    plt.scatter(region_data['long'],
                region_data['lat'],
                label=region,
                edgecolors='black')

# Adicionando rótulos
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title('Regiões do Brasil')
ax.legend()

# Show the plot
plt.show()
```

👉 Essas são as regiões do Brasil.

* Os círculos representam a localização das capitais de cada Estado.

👉 Para criar o agrupamento, utilizaremos as ferramentas do pacote `sklearn`

```python
# Carregando pacote e partes necessarias
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
import numpy as np

# Criando um DataFrame com as informações que serão utilizadas como métricas
df_regioes = brasil[['long',	'lat']]
df_regioes.index = brasil['name_state']
df_regioes
```

> ⚠️ São permitidas somente variáveis numéricas para que sejam realizados os cálculos.

👉 Em `df_regioes`, estão contidas a `longitude` e `latitude` das capitais (26 Estados e Distrito Federal)

```python
# Interligando passos
pipeline = make_pipeline(
    StandardScaler(),          # Padronizando dados
    KMeans(n_clusters=5,       # Num de grupos
           random_state = 321, # Semente aleatória
           n_init = "auto")    # Mantenha essa opção no "auto"
)

# Agrupamento
grupos = pipeline.fit(df_regioes) # DataFrame com as métricas

# Exibindo resultado
print(grupos.predict(df_regioes))
```

> ⚠️ **ANALISANDO CADA ITEM**

`n_clusters` = Número de grupos (*cluster* = grupo em português)

`random_state` = É um número qualquer que utilizamos para que o algoritmo se inicie sempre da mesma forma. Pode ser o valor que você preferir, aqui utilizamos `123`. Tente outro valor. Possivelmente, o resultado será diferente do obtido.

`grupos.predict()` = É a previsão/rótulo de cada grupo predito para cada observação.

> ⚠️ **INTERPRETANDO**

Este método simplesmente aloca as informações na quantidade de grupos pré determinada em `n_clusters`.

👉 Cabe a nós interpretar o que significa cada grupo.

```python
# Juntando a regiao calculada com a localização
df_regioes['regiao'] = grupos.predict(df_regioes)
df_regioes
```

👉 Com a classificação criada, podemos visualizar como este método agruparia os Estados do Brasil em 5 regiões:

```python
# Criando figura
fig, ax = plt.subplots()

# Lista de cores
region_colors = ['tab:blue', 'tab:orange', 'tab:green',
                 'tab:red', 'tab:purple']

# Desenhando cada região
regions = brasil['region'].unique()

for i, region in enumerate(regions):
    region_data = brasil.loc[brasil['region'] == region, :]
    region_data.plot(ax=ax, edgecolor='black',
                     facecolor=region_colors[i],
                     linewidth=0.5,
                     label=region)

for i, region in enumerate(regions):
    class_data = df_regioes.loc[df_regioes['regiao'] == i, :]
    plt.scatter(class_data['long'],
                class_data['lat'],
                marker=f'${i}$',
                edgecolors='black')

# Adicionando rótulos
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title('Agrupamento automático das Regiões do Brasil')
ax.legend()

# Show the plot
plt.show()
```

> ⚠️ **ATENÇÃO**

* Os nomes do grupos (neste caso, representado por números de 0 à 4) são permutáveis. Ou seja, o grupo `0` pode ser chamado de `1` e vice-versa toda vez que alterarmos o argumento `random_state`.

* Não leve tão a sério este exemplo! Estamos exclusivamente considerando a localização das capitais. Sabemos que as regiões foram formadas considerando outros aspectos políticos e econômicos.

> 🛠️ *Exemplo prático 1:* Agrupando e classificando ações

> _O *retorno percentual* pode ser utilizado como um indicador de valorização de uma ação. Por outro lado, o *volume* pode ser visto como um indicador de liquidez (ou seja, é mais fácil comprar e vender ações sem afetar significativamente o preço de mercado)_

👉 Utilizando estes 2 indicadores, vamos agrupar 23 ações em 2 grupos considerando o 1o trimestre de 2023

```python
# Carregando pacote
import yfinance as yf

# Tickers
tickers = ["ABEV3.SA", "AMER3.SA", "ASAI3.SA", "B3SA3.SA", "BBDC4.SA",
           "BRFS3.SA", "CIEL3.SA", "COGN3.SA", "CPLE6.SA", "CSNA3.SA",
           "CVCB3.SA", "HAPV3.SA", "IFCM3.SA", "ITSA4.SA", "ITUB4.SA",
           "MGLU3.SA", "PETR4.SA", "RAIL3.SA", "SEQL3.SA", "SOMA3.SA",
           "USIM5.SA", "VALE3.SA", "VIIA3.SA"]

# Obtendo dados
df_retorno = yf.download(tickers,                # Ação
                         start="2023-01-01",     # Data inicial AAAA-MM-DD
                         end="2023-04-01",       # Data final AAAA-MM-DD
                         progress=False).iloc[:, list(range(23))]
                         # Selecionando as 23 primeiras colunas

df_volume = yf.download(tickers,                # Ação
                        start="2023-01-01",     # Data inicial AAAA-MM-DD
                        end="2023-04-01",       # Data final AAAA-MM-DD
                        progress=False).iloc[:, list(range(-23, 0))]
                        # Selecionando as 23 ultimas colunas

# Renomeando colunas
df_retorno.columns = tickers
df_volume.columns = tickers
```

```python
# Visualizando resultado
df_retorno
```

```python
# Visualizando resultado
df_volume
```

👉 Precisamos calcular o `retorno percentual = valor atual / valor anterior * 100`. Utilzaremos o método `.pct_change()`

```python
# Calculando retorno percentual
# > Para cada colunas de `df_retorno`,
# > substituo o valor pelo retorno percentual
for col in df_retorno.columns:
  df_retorno[col] = df_retorno[col].pct_change() * 100

# Visualizando resultado
df_retorno
```

👉 A primeira linha é inviabilizada por conta deste cálculo. Vamos removê-la nos 2 `DataFrames`

```python
# Removendo a primeira linha em ambos
df_retorno = df_retorno.iloc[1:]
df_volume = df_volume.iloc[1:]
```

👉 Precisamos de um único valor para representar todo o trimestre. Vamos utilizar a **média** para resumir toda a informação

```python
# Crie um novo DataFrame com a média das colunas
df_indicadores = pd.DataFrame([df_retorno.mean(),
                               df_volume.mean() / 100000], # Divisão por 100.000
                                                           # para melhor visualização
                              index = ['Retorno', 'Volume'],
                              columns=tickers)

# Invertendo linhas por colunas
df_indicadores = df_indicadores.transpose()

# Visualizando resultado
df_indicadores
```

👉 Da mesma forma que fizemos no exemplo introdutório, vamos compreender estes 2 indicadores como 'localizações' em um mapa

```python
# Carregando pacotes
import matplotlib.pyplot as plt

# Gráfico de dispersão
plt.scatter(df_indicadores['Retorno'], df_indicadores['Volume'], c='black')

# Adicionando rótulos aos pontos
for i, txt in enumerate(tickers):
    plt.annotate(txt, (df_indicadores['Retorno'][i], df_indicadores['Volume'][i]),
                 fontsize=8, ha='left', c='grey')

# Definindo rótulos dos eixos
plt.xlabel('Retorno')
plt.ylabel('Volume')

# Definindo o título do gráfico
plt.title('Retorno vs. Volume')

# Exibindo o gráfico
plt.grid(True)
plt.show()
```

👉 Finalmente, vamos criar 2 grupos para estas ações

```python
# Carregando pacote e partes necessarias
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
import numpy as np

# Interligando passos
pipeline = make_pipeline(
    StandardScaler(),          # Padronizando dados
    KMeans(n_clusters=2,       # Num de grupos
           random_state = 999, # Semente aleatória (pode usar a que você quiser)
           n_init = "auto")    # Mantenha essa opção no "auto"
)

# Agrupamento
grupos_acoes = pipeline.fit(df_indicadores) # DataFrame com as métricas

# Juntando o grupo calculado com os indicadores
df_indicadores['Grupo'] = grupos_acoes.predict(df_indicadores)
df_indicadores
```

👉 Vamos visualizar o resultado

```python
# Gráfico de dispersão
plt.scatter(df_indicadores['Retorno'], df_indicadores['Volume'], c='white')

# Lista de cores
cores_grupo = ['tab:blue', 'tab:orange']

# Adicionando rótulos aos pontos
for i, txt in enumerate(tickers):
    plt.annotate(txt, (df_indicadores['Retorno'][i], df_indicadores['Volume'][i]),
                 fontsize=8, ha='left', c='grey')
    plt.scatter(df_indicadores['Retorno'][i],
                df_indicadores['Volume'][i],
                c=cores_grupo[df_indicadores['Grupo'][i]])

# Definindo rótulos dos eixos
plt.xlabel('Retorno')
plt.ylabel('Volume')

# Definindo o título do gráfico
plt.title('Retorno vs. Volume')

# Exibindo o gráfico
plt.grid(True)
plt.show()
```

👉 Vamos replicar agora para 4 grupos e visualizar o resultado

```python
# Interligando passos
pipeline = make_pipeline(
    StandardScaler(),          # Padronizando dados
    KMeans(n_clusters=4,       # Num de grupos
           random_state = 999, # Semente aleatória (pode usar a que você quiser)
           n_init = "auto")    # Mantenha essa opção no "auto"
)

# Agrupamento
grupos_acoes = pipeline.fit(df_indicadores) # DataFrame com as métricas

# Juntando o grupo calculado com os indicadores
df_indicadores['Grupo'] = grupos_acoes.predict(df_indicadores)
df_indicadores

# Gráfico de dispersão
plt.scatter(df_indicadores['Retorno'], df_indicadores['Volume'], c='white')

# Lista de cores
cores_grupo = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red']

# Adicionando rótulos aos pontos
for i, txt in enumerate(tickers):
    plt.annotate(txt, (df_indicadores['Retorno'][i], df_indicadores['Volume'][i]),
                 fontsize=8, ha='left', c='grey')
    plt.scatter(df_indicadores['Retorno'][i],
                df_indicadores['Volume'][i],
                c=cores_grupo[df_indicadores['Grupo'][i]])

# Definindo rótulos dos eixos
plt.xlabel('Retorno')
plt.ylabel('Volume')

# Definindo o título do gráfico
plt.title('Retorno vs. Volume')

# Exibindo o gráfico
plt.grid(True)
plt.show()
```

Agora, cabe a você interpretar o que cada grupo significa.

> 🛠️ *Exemplo prático 2:* Agrupando e classificando países

> *O **Produto Interno Bruto (PIB) per capita** e a **taxa de desemprego** são duas métricas econômicas importantes que desempenham papéis distintos na avaliação e no entendimento da saúde econômica de um país.*
>
> * Produto Interno Bruto (PIB) per capita: https://data.worldbank.org/indicator/NY.GDP.PCAP.CD?view=chart
>
> * Taxa de desemprego: https://data.worldbank.org/indicator/SL.UEM.TOTL.ZS?view=chart

> 👉 Vamos utilizar estes indicadores (2021) para classificar países

```python
# Carregando módulo (parte do pacote)
from pandas_datareader import wb

siglas_paises = [
    'CA',  # Canadá
    'MX',  # México
    'GB',  # Reino Unido
    'FR',  # França
    'DE',  # Alemanha
    'IT',  # Itália
    'ES',  # Espanha
    'PT',  # Portugal
    'AU',  # Austrália
    'NZ',  # Nova Zelândia
    'JP',  # Japão
    'CN',  # China
    'IN',  # Índia
    'BR',  # Brasil
    'AR',  # Argentina
    'CL',  # Chile
    'CO',  # Colômbia
    'PE',  # Peru
    'EG',  # Egito
    'ZA',  # África do Sul
    'NG',  # Nigéria
    'KE',  # Quênia
    'RU',  # Rússia
    'TR',  # Turquia
    'SA',  # Arábia Saudita
    'AE',  # Emirados Árabes Unidos
    'KR',  # Coreia do Sul
    'US'   # Estados Unidos
]

# Obtendo dados
df_wb = wb.download(indicator = ['NY.GDP.PCAP.CD',    # PIB per capita (US$)
                                 'SL.UEM.TOTL.ZS'],   # Desemprego (%)
                    country = siglas_paises,          # Lista de países
                    start = 2021,                     # Data início AAAA
                    end = 2021)                       # Data fim AAAA

# Editando DataFrame
df_wb = df_wb.reset_index()
df_wb.index = df_wb['country']
df_wb = df_wb.loc[:, ['NY.GDP.PCAP.CD', 'SL.UEM.TOTL.ZS']]
df_wb.columns = ['pib', 'desemprego']
df_wb['pib'] = df_wb['pib'] / 1000

# Resultado (tabela)
df_wb
```

👉 Vamos visualizar como os países se distribuem a partir destas métricas

```python
# Gráfico de dispersão
plt.scatter(df_wb['pib'], df_wb['desemprego'], c='black')

# Adicionando rótulos aos pontos
for i, txt in enumerate(df_wb.index):
    plt.annotate(txt, (df_wb['pib'][i], df_wb['desemprego'][i]),
                 fontsize=8, ha='left', c='grey')

# Definindo rótulos dos eixos
plt.xlabel('PIB per capita (US$)')
plt.ylabel('Desemprego (%)')

# Exibindo o gráfico
plt.grid(True)
plt.show()
```

👉 Vamos agrupar os países em 5 *clusters*

```python
# Interligando passos
pipeline = make_pipeline(
    StandardScaler(),          # Padronizando dados
    KMeans(n_clusters=5,       # Num de grupos
           random_state = 777, # Semente aleatória (pode usar a que você quiser)
           n_init = "auto")    # Mantenha essa opção no "auto"
)

# Agrupamento
grupos_paises = pipeline.fit(df_wb) # DataFrame com as métricas

# Juntando o grupo calculado com os indicadores
df_wb['grupo'] = grupos_paises.predict(df_wb)
df_wb

# Gráfico de dispersão
plt.scatter(df_wb['pib'], df_wb['desemprego'], c='black')

# Lista de cores
cores_grupo = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red',
               'tab:purple']

# Adicionando rótulos aos pontos
for i, txt in enumerate(df_wb.index):
    plt.annotate(txt, (df_wb['pib'][i], df_wb['desemprego'][i]),
                 fontsize=8, ha='left', c='grey')
    plt.scatter(df_wb['pib'][i],
                df_wb['desemprego'][i],
                c=cores_grupo[df_wb['grupo'][i]])

# Definindo rótulos dos eixos
plt.xlabel('PIB per capita (US$)')
plt.ylabel('Desemprego (%)')

# Exibindo o gráfico
plt.grid(True)
plt.show()
```
