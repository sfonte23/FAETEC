## Estat√≠stica Descritiva no Mercado Financeiro

üìã A estat√≠stica descritiva desempenha um papel fundamental nas an√°lises do mercado financeiro, sendo uma ferramenta indispens√°vel para investidores, analistas e tomadores de decis√£o. Essa disciplina estat√≠stica permite que os profissionais financeiros compreendam, interpretem e comuniquem dados financeiros de maneira eficaz, fornecendo insights valiosos para a tomada de decis√µes informadas. Aqui est√£o algumas das raz√µes pelas quais a estat√≠stica descritiva √© importante no contexto do mercado financeiro:


1.	**Resumo e Organiza√ß√£o de Dados**: A estat√≠stica descritiva ajuda a resumir grandes conjuntos de dados financeiros, tornando-os mais compreens√≠veis. Ela permite a organiza√ß√£o dos dados em tabelas, gr√°ficos e medidas de resumo, como m√©dias, medianas e desvios padr√£o.

2.	**Identifica√ß√£o de Tend√™ncias e Padr√µes**: Ao analisar s√©ries temporais de pre√ßos de ativos ou indicadores econ√¥micos, a estat√≠stica descritiva ajuda a identificar tend√™ncias, sazonalidades e padr√µes que podem ser usados para tomar decis√µes de investimento informadas.

3.	**Medi√ß√£o de Risco**: A estat√≠stica descritiva √© essencial para a avalia√ß√£o de risco no mercado financeiro. O c√°lculo de medidas de volatilidade, como o desvio padr√£o, ajuda os investidores a entender o grau de varia√ß√£o nos pre√ßos dos ativos e a gerenciar seus riscos.

4.	**Compara√ß√£o de Investimentos**: Investidores frequentemente precisam comparar o desempenho de diferentes ativos financeiros, como a√ß√µes, t√≠tulos ou fundos m√∫tuos. A estat√≠stica descritiva fornece ferramentas para realizar essas compara√ß√µes de maneira objetiva.

5.	**Tomada de Decis√µes Informadas**: Com base nas an√°lises estat√≠sticas descritivas, os investidores podem tomar decis√µes informadas sobre a aloca√ß√£o de ativos, a constru√ß√£o de portf√≥lios e a sele√ß√£o de estrat√©gias de investimento.

6.	**Avalia√ß√£o de Desempenho de Fundos e Gestores**: Para avaliar o desempenho de fundos de investimento e gestores de carteira, a estat√≠stica descritiva √© usada para calcular indicadores que medem o retorno ajustado pelo risco.

7.	**Compreens√£o da Distribui√ß√£o de Retornos**: A estat√≠stica descritiva ajuda a entender a distribui√ß√£o de retornos de ativos financeiros, o que √© crucial para a modelagem de riscos e a elabora√ß√£o de estrat√©gias de gerenciamento de portf√≥lio.

8.	**Monitoramento de Eventos Econ√¥micos**: A an√°lise de dados econ√¥micos, como taxas de juros, infla√ß√£o e indicadores de mercado de trabalho, por meio de t√©cnicas estat√≠sticas descritivas, permite aos investidores antecipar eventos econ√¥micos e suas implica√ß√µes nos mercados financeiros.

9.	**Relat√≥rios Financeiros**: Empresas e institui√ß√µes financeiras usam estat√≠sticas descritivas para apresentar seus resultados financeiros de forma clara e concisa em relat√≥rios, tornando-os acess√≠veis aos investidores e reguladores.

10.	**Avalia√ß√£o de Riscos de Cr√©dito**: A estat√≠stica descritiva √© crucial para avaliar o risco de cr√©dito ao analisar o hist√≥rico de pagamentos de devedores e calcular m√©tricas como o rating de cr√©dito.

> Em resumo, a estat√≠stica descritiva desempenha um papel essencial nas an√°lises do mercado financeiro, fornecendo as ferramentas necess√°rias para resumir, analisar e interpretar dados financeiros. Ela ajuda os investidores e analistas a compreenderem os mercados, identificar tend√™ncias, gerenciar riscos e avaliar o desempenho de investimentos. Sem ela, a tomada de decis√µes informadas no mundo financeiro seria muito mais desafiadora e arriscada. ## Medidas de posi√ß√£o

üîõ As medidas de posi√ß√£o s√£o ferramentas estat√≠sticas fundamentais que nos ajudam a entender a distribui√ß√£o dos dados em um conjunto de informa√ß√µes. Elas fornecem *insights* valiosos sobre onde os valores est√£o centralizados.

### M√©dia

üî£ √â a medida de posi√ß√£o mais conhecida e amplamente usada. Ela √© calculada somando todos os valores de um conjunto de dados e dividindo o resultado pelo n√∫mero de observa√ß√µes:

```python
# Carregando pacotes
import pandas as pd

# Dados ficticios
dados = pd.Series([8, 6, 5, 7, 3, 2])

# Calculando a m√©dia manualmente:
# 1) somando todos os valores de um conjunto de dados
#    8 + 6 + 5 + 7 + 3 + 2
# 2) dividindo o resultado pelo n√∫mero de observa√ß√µes
#    / 6
media = (8 + 6 + 5 + 7 + 3 + 2) / 6
print(f'M√©dia: {media}')

# Calculando a m√©dia de um objeto pd.Series ou pd.DataFrame
print(f'M√©dia: {dados.mean()}')
```

### M√©dia m√≥vel

üî£ √â uma ferramenta valiosa para suavizar dados, identificar tend√™ncias, fazer previs√µes e tomar decis√µes informadas em uma variedade de campos, especialmente quando se lida com s√©ries temporais ou conjuntos de dados que exibem varia√ß√µes ao longo do tempo.

> üõ†Ô∏è *Exemplo pr√°tico 2:* Calculando m√©dias m√≥veis para determinadas a√ß√µes
>  * Calcular medidas de dispers√£o por um per√≠odo muito longo de tempo pode levar a √≠ndices enviesados

```python
# Definindo abrevia√ß√£o (ticker)
ticker = ["BBAS3.SA", # Banco do Brasil
          "BBDC4.SA", # Bradesco
          "ITUB4.SA"] # Itau

# Obtendo dados
df_acao = yf.download(ticker,                # A√ß√£o
                      start="2014-01-01",    # Data inicial AAAA-MM-DD
                      end="2023-04-01",      # Data final AAAA-MM-DD
                      progress=False)

# Selecionando a coluna `Adj Close` pela posi√ß√£o
df_acao = df_acao.iloc[:, [0, 1, 2]]

# Renomeando coluna
df_acao.columns = ticker

# Visualizando resultado
df_acao
```

üëâ Calculando a m√©dia m√≥vel das a√ß√µes
*   √â necess√°rio definir um p√©riodo retroativo para se calcular a m√©dia m√≥vel.
*   Como estamos tratando de dados di√°rios, podemos considerar 5 (5 dias √∫teis = 1 semana, por exemplo)

```python
# Iniciando um DataFrame vazio
media_movel = pd.DataFrame()

# Calcular a m√©dia m√≥vel de 5 per√≠odos
media_movel['BBAS3.SA_MM'] = df_acao['BBAS3.SA'].rolling(5).mean()
media_movel['BBDC4.SA_MM'] = df_acao['BBDC4.SA'].rolling(5).mean()
media_movel['ITUB4.SA_MM'] = df_acao['ITUB4.SA'].rolling(5).mean()
media_movel
```

üí¨ Note que os primeiro per√≠odos possuem `NaN`. Ou seja, n√£o temos informa√ß√£o pois utilizamos estes valores para calcular as m√©dias a frente!

```python
# Criando o gr√°fico
df_acao.plot(color=['goldenrod', 'firebrick', 'coral'])

# Adicionando:
plt.title('Valores de fechamento: 2014/T1 a 2023/T1') # T√≠tulo
plt.xlabel('Data')                 # R√≥tulo do eixo horizontal (x)
plt.ylabel('Valor (R$)')           # R√≥tulo do eixo vertical (y)

# Adicionando as m√©dias m√≥veis
plt.plot(media_movel.index.to_numpy(),
         media_movel['BBAS3.SA_MM'],
         color='black',  # Cor = preto
         linewidth=.5,   # Largura reduzida
         linestyle='dashed') # Linha tracejada

plt.plot(media_movel.index.to_numpy(),
         media_movel['BBDC4.SA_MM'],
         color='black',  # Cor = preto
         linewidth=.5,   # Largura reduzida
         linestyle='dashed') # Linha tracejada

plt.plot(media_movel.index.to_numpy(),
         media_movel['ITUB4.SA_MM'],
         color='black',  # Cor = preto
         linewidth=.5,   # Largura reduzida
         linestyle='dashed') # Linha tracejada

# Exibindo o gr√°fico
plt.show()
```

üí¨ Note que as m√©dias m√≥veis s√£o mais suaves e cont√©m menos movimentos bruscos.

*    üëÅÔ∏è‚Äçüó®Ô∏è A partir deste m√©todo, j√° conseguimos fazer previs√µes: A partir das √∫ltimas 5 cota√ß√µes, podemos predizer o valor da a√ß√£o para o pr√≥ximo dia!

```python
# Recuperando os 5 √∫ltimos valores
df_acao['BBAS3.SA'].tail(5)
```

```python
# Uma poss√≠vel previs√£o para o pr√≥ximo dia √∫til (03/04/2023) seria:
previsao_mm = df_acao['BBAS3.SA'].tail(5).mean()
print(f'BBAS3.SA | Previs√£o para 03/04/2023: R${round(previsao_mm, 2)}')
```

## Medidas de dispers√£o

üîõ Medidas de dispers√£o s√£o componentes cruciais que nos ajudam a compreender a extens√£o com a qual os valores em um conjunto de dados se afastam da medida de tend√™ncia central, como a m√©dia. Elas fornecem *insights* importantes sobre a variabilidade e a distribui√ß√£o dos dados. As medidas de dispers√£o mais comuns s√£o:

### Vari√¢ncia

üî£ √â uma medida que quantifica a dispers√£o dos dados em rela√ß√£o √† m√©dia. Ela √© calculada ao encontrar a m√©dia dos quadrados das diferen√ßas entre cada valor e a m√©dia do conjunto de dados:

```python
# Dados ficticios
dados = pd.Series([8, 6, 5, 7, 3, 2])

# Calculando a vari√¢ncia de um objeto pd.Series ou pd.DataFrame
print(f'Vari√¢ncia: {dados.var()}')
```

> ‚ö†Ô∏è **ALGUMAS OBSERVA√á√ïES SOBRE A VARI√ÇNCIA**
>
> A vari√¢ncia:
> * alta indica que os valores do conjunto de dados est√£o amplamente dispersos em rela√ß√£o √† m√©dia
> * baixa sugere que os valores est√£o mais pr√≥ximos da m√©dia.
> * nula implica que os valores s√£o todos iguais.
> * √© sempre maior ou igual √† zero.
> * √© uma medida com escala quadr√°tica. Se os dados se referem √† valores (R\$), a vari√¢ncia est√° em uma medida de valores ao quadrado (R\$ $^2$), dif√≠cil de interpretar.

### Desvio-Padr√£o

üî£ √â a raiz quadrada da vari√¢ncia e √© uma medida de dispers√£o mais utilizada e interpret√°vel pois tem as mesmas unidades originais dos dados, o que o torna mais intuitivo para a compreens√£o da dispers√£o.

```python
# Dados ficticios
dados = pd.Series([8, 6, 5, 7, 3, 2])

# Calculando o desvio-padr√£o de um objeto pd.Series ou pd.DataFrame
print(f'Desvio-padr√£o: {dados.std()}')

# Demonstrando que desvio padrao = raiz quadrada da variancia
print(f'Desvio-padr√£o: {dados.var() ** (1/2)}')
```

> ‚ö†Ô∏è Quanto maior o desvio-padr√£o, maior a dispers√£o dos dados em rela√ß√£o √† m√©dia, e vice-versa (assim como na vari√¢ncia).

## Volatilidade

„ÄΩÔ∏è A volatilidade no mercado financeiro se refere √† medida da varia√ß√£o dos pre√ßos de ativos financeiros, como a√ß√µes, t√≠tulos, moedas e commodities, ao longo do tempo. Ela descreve a magnitude das flutua√ß√µes de pre√ßo e a incerteza associada aos movimentos desses ativos. A volatilidade √© uma parte intr√≠nseca dos mercados financeiros e desempenha um papel crucial na tomada de decis√µes de investimento e gest√£o de riscos. Alguns pontos importantes sobre a volatilidade:

1.   **Natureza Inerente**: A volatilidade √© uma caracter√≠stica natural dos mercados financeiros. Os pre√ßos dos ativos podem ser influenciados por uma ampla gama de fatores, como not√≠cias econ√¥micas, eventos pol√≠ticos, mudan√ßas nas taxas de juros, condi√ß√µes macroecon√¥micas e sentimentos do mercado. Esses fatores, por sua natureza imprevis√≠vel, contribuem para a volatilidade dos pre√ßos.

2.   **Medi√ß√£o**: A volatilidade √© frequentemente medida usando indicadores estat√≠sticos, sendo o mais comum deles o desvio-padr√£o ou a vari√¢ncia dos retornos de um ativo. Quanto maior for o desvio-padr√£o, maior a volatilidade.

3.   **Impacto nos Investimentos**: A volatilidade afeta diretamente a rentabilidade e o risco dos investimentos. Ativos mais vol√°teis t√™m o potencial de gerar retornos significativamente mais altos, mas tamb√©m apresentam riscos substanciais de perdas. Investidores geralmente consideram a volatilidade ao escolher ativos para suas carteiras e ao ajustar a aloca√ß√£o de ativos de acordo com seus objetivos e toler√¢ncia ao risco.

4.    **Estrat√©gias de Negocia√ß√£o**: *Traders* e investidores utilizam estrat√©gias espec√≠ficas para aproveitar a volatilidade do mercado. Por exemplo, o *day trading* √© uma abordagem em que os *traders* aproveitam as flutua√ß√µes de curto prazo para lucrar. Da mesma forma, existem produtos financeiros, como op√ß√µes e contratos futuros, projetados para ajudar os investidores a proteger ou alavancar suas posi√ß√µes em resposta √† volatilidade.

5.    **Gest√£o de Riscos**: A volatilidade √© um componente cr√≠tico na gest√£o de riscos financeiros. Empresas, investidores institucionais e indiv√≠duos usam estrat√©gias de gerenciamento de riscos, como a diversifica√ß√£o de carteiras e o uso de derivativos, para proteger seus investimentos contra movimentos de pre√ßos adversos.

6.    **Indicador Econ√¥mico**: A volatilidade tamb√©m √© usada como um indicador econ√¥mico. Em geral, mercados financeiros mais vol√°teis podem refletir incertezas na economia global. Os analistas observam a volatilidade como um sinal de poss√≠veis mudan√ßas no ambiente econ√¥mico e pol√≠tico.

> Em resumo, a volatilidade √© uma caracter√≠stica inerente aos mercados financeiros que descreve a magnitude das flutua√ß√µes de pre√ßo ao longo do tempo. Ela √© fundamental para a tomada de decis√µes de investimento, a gest√£o de riscos e a compreens√£o das condi√ß√µes econ√¥micas e de mercado. Investidores e traders precisam considerar a volatilidade ao tomar decis√µes informadas sobre como alocar seus recursos e quais estrat√©gias adotar.

> üõ†Ô∏è *Exemplo pr√°tico 3:* Calculando a volatilidade para diferentes moedas

```python
# Definindo abrevia√ß√£o (ticker)
ticker = ["BRL=X", # Real (Brasil)
          "CAD=X", # D√≥lar canadense (Canad√°)
          "CHF=X", # Franco su√≠√ßo (Su√≠√ßa)
          "GBPUSD=X"] # Libra esterlina (Reino Unido)

# Obtendo dados
df_cambio = yf.download(ticker,                # Moedas
                        start="2023-01-01",    # Data inicial AAAA-MM-DD
                        end="2023-04-01",      # Data final AAAA-MM-DD
                        progress=False)

# Selecionando a coluna `Adj Close` pela posi√ß√£o
df_cambio = df_cambio.iloc[:, [0, 1, 2, 3]]

# Renomeando coluna
df_cambio.columns = ["BRL", "CAD", "CHF", "GBP"]

# Visualizando resultado
df_cambio
```

```python
# Criando o gr√°fico
df_cambio.plot(color=['darkgreen', 'tomato',
                      'red', 'royalblue'])

# Adicionando:
plt.title('C√¢mbio')
plt.xlabel('Data')
plt.ylabel('Valor (moeda corrente)')

# Exibindo o gr√°fico
plt.show()
```

## Correla√ß√£o linear

üîÄ √â uma medida estat√≠stica que descreve o grau de relacionamento linear entre duas vari√°veis em um conjunto de dados. Ela √© usada para determinar se existe uma rela√ß√£o linear entre duas vari√°veis e, em caso afirmativo, qual √© a for√ßa e a dire√ß√£o dessa rela√ß√£o.

A correla√ß√£o linear √© frequentemente representada pelo coeficiente de correla√ß√£o (tamb√©m chamado de $r$) varia de $-1$ a $1$ e fornece as seguintes informa√ß√µes:

*   **Correla√ß√£o Negativa ($r < 0$)**: Quando o valor de $r$ est√° pr√≥ximo de -1, indica uma correla√ß√£o negativa forte entre as vari√°veis. Isso significa que, √† medida que uma vari√°vel aumenta, a outra tende a diminuir.

*    **Aus√™ncia de Correla√ß√£o ($r = 0$)**: Quando o valor de $r$ est√° pr√≥ximo de zero, indica que n√£o h√° uma rela√ß√£o linear aparente entre as vari√°veis. No entanto, √© importante notar que a aus√™ncia de correla√ß√£o linear n√£o implica necessariamente na aus√™ncia de outras rela√ß√µes n√£o lineares.

> ‚ö†Ô∏è **A correla√ß√£o linear s√≥ mede relacionamentos lineares entre vari√°veis**. Se houver uma rela√ß√£o n√£o linear entre as vari√°veis, a correla√ß√£o linear n√£o ser√° uma medida apropriada para avaliar essa rela√ß√£o.
