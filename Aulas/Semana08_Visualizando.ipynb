## Importância das Ferramentas Gráficas

🔭 Esta ferramenta oferece uma forma poderosa de compreender, analisar e comunicar informações complexas de maneira eficaz. Nesse contexto, a visualização de dados se refere à representação gráfica e interativa de informações financeiras por meio de gráficos, tabelas, mapas e outras representações visuais. Aqui estão algumas razões pelas quais a visualização de dados é tão importante no mercado financeiro:

1.	**Compreensão Rápida e Intuitiva**: O mercado financeiro lida com uma grande quantidade de dados, incluindo preços de ativos, indicadores econômicos, taxas de juros, volumes de negociação e muito mais. A visualização de dados permite que os profissionais financeiros absorvam informações complexas de maneira mais rápida e intuitiva do que por meio de números em tabelas.

Ao observar visualmente como os preços dos ativos estão se movendo ao longo do tempo, os investidores podem detectar tendências de alta, baixa ou lateralização.

3.	**Análise Comparativa**: A visualização de dados permite a comparação de múltiplos ativos, indicadores ou métricas de uma só vez. Isso ajuda os analistas financeiros a entenderem como diferentes ativos estão se comportando em relação uns aos outros e como diferentes fatores econômicos estão afetando o mercado como um todo.

4.	**Decisões de Investimento Informadas**: Investidores e gestores de fundos usam visualizações de dados para avaliar o desempenho de suas carteiras de investimento, identificar riscos e oportunidades, e ajustar suas estratégias com base em *insights* visuais. Isso ajuda a tomar decisões mais informadas e a mitigar riscos desnecessários.

5.	**Acompanhamento do Mercado em Tempo Real**: Com as tecnologias atuais, é possível criar visualizações de dados em tempo real que refletem as mudanças no mercado instantaneamente. Isso é crucial para os traders e investidores que precisam reagir rapidamente a eventos que podem afetar os mercados.

6.	**Comunicação Eficaz**: As visualizações de dados não são apenas valiosas para os profissionais financeiros, mas também para a comunicação com partes interessadas. Ao apresentar informações complexas de maneira visual, as empresas financeiras podem explicar estratégias, desempenho e perspectivas de maneira mais compreensível para investidores, clientes e parceiros.

7.	**Monitoramento de Riscos**: As visualizações de dados ajudam na identificação e monitoramento de riscos financeiros. Ao visualizar métricas como volatilidade, exposição a determinados setores ou regiões, e correlações entre ativos, os gestores de risco podem tomar medidas para mitigar possíveis impactos negativos.
Em resumo, a visualização de dados desempenha um papel fundamental na análise, tomada de decisões e comunicação no mercado financeiro. Ela transforma dados complexos em informações acessíveis, ajudando os profissionais financeiros a ganhar *insights* valiosos e tomar ações mais informadas.

## Boas práticas na criação de gráficos

🎨 A criação de gráficos eficazes é um processo que requer uma compreensão profunda tanto dos dados subjacentes quanto dos princípios de design visual. Ao unir esses dois elementos, os profissionais podem criar representações visuais que não apenas são atraentes esteticamente, mas também são precisas, claras e informativas.

1. **Escolha do Tipo de Gráfico Adequado**: O primeiro passo crucial é selecionar o tipo de gráfico mais apropriado para os dados que você deseja apresentar. Diferentes tipos de gráficos, como gráficos de barras, gráficos de linhas, gráficos de pizza e gráficos de dispersão, são adequados para diferentes tipos de dados e mensagens. A escolha certa maximiza a compreensão e minimiza a ambiguidade.

2. **Simplicidade e Clareza**: Menos é mais quando se trata de criar gráficos. Evite o excesso de informações, como rótulos confusos, muitos marcadores ou muitas cores. Priorize a simplicidade e a clareza, garantindo que os elementos visuais contribuam diretamente para a compreensão dos dados, sem sobrecarregar o espectador.

3. **Legibilidade dos Dados**: Certifique-se de que os valores exibidos nos gráficos sejam legíveis. Isso inclui a escolha de fontes adequadas, tamanhos de fonte apropriados e rótulos de eixo descritivos. A legibilidade é fundamental para garantir que o público possa interpretar os dados com precisão.

4. **Contextualização**: Sempre forneça contexto aos dados apresentados. Títulos claros, legendas explicativas e rótulos de eixo bem elaborados ajudam a situar os espectadores no que estão vendo, garantindo uma interpretação correta.

5. **Coerência Visual**: Mantenha uma coerência visual em todo o gráfico e entre diferentes gráficos em uma apresentação. Isso envolve a escolha consistente de paletas de cores, estilos de linhas e formatos de marcadores, de modo que a visualização seja unificada e fácil de seguir.

6. **Destaque de Informações Cruciais**: Se houver informações ou pontos específicos que você deseja destacar, use realces, anotações ou elementos visuais para chamar a atenção do público. Esses destaques podem direcionar o foco para aspectos importantes dos dados.

7. **Minimização de Distrações**: Evite elementos visuais que possam distrair do objetivo principal do gráfico. Isso inclui efeitos visuais exagerados, cores desnecessárias ou elementos decorativos que não contribuam para a compreensão dos dados.

8. **Título Explicativo**: Um título claro e conciso ajuda a transmitir a mensagem geral do gráfico. Ele deve resumir o que está sendo mostrado e destacar os principais insights.

9. **Atualização e Manutenção**: Se os dados subjacentes mudarem com o tempo, certifique-se de atualizar seus gráficos regularmente para refletir as mudanças. Além disso, periodicamente revise e otimize seus gráficos para garantir que permaneçam eficazes e relevantes.

10. **Teste e Feedback**: Antes de apresentar seus gráficos a um público mais amplo, peça feedback de colegas ou especialistas. Isso pode ajudar a identificar possíveis problemas de interpretação, legibilidade ou formatação.

> 👨‍🎨👩‍🎨 Em última análise, a criação de gráficos eficazes é uma combinação de arte e ciência, exigindo sensibilidade visual e habilidades analíticas. Ao aderir a boas práticas e investir tempo na criação e revisão de gráficos, você pode garantir que suas representações visuais transmitam informações com precisão, clareza e impacto.

## Criando gráficos

🎨 Vamos utilizar o pacote `matplotlib`: biblioteca essencial para a visualização de dados em Python, oferecendo uma ampla variedade de recursos e flexibilidade para criar gráficos de alta qualidade. É uma ferramenta valiosa para comunicar informações complexas de forma clara e eficaz.

### Gráficos de linhas

📉 Gráficos de linhas são uma ferramenta valiosa para analisar e comunicar dados, particularmente quando se trata de representar tendências e mudanças ao longo do tempo.

> 🛠️ *Exemplo prático 1:* Visualizando valores de uma determinada ação ao longo do tempo

```python
# Carregando pacotes
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Definindo abreviação (ticker)
ticker = ["PETR4.SA"]  # Petrobras SA

# Obtendo dados
df_acao = yf.download(ticker,                # Ação
                      start="2023-01-01",    # Data inicial AAAA-MM-DD
                      end="2023-04-01")      # Data final AAAA-MM-DD

# Resultado (tabela)
df_acao
```

👉 Relembrando que só temos interesse na coluna `Adj Close`
*   Vamos selecionar somente esta coluna (e manter as datas)
*   Vamos renomear para `PETR4.SA`

```python
# Selecionando a coluna `Adj Close` pelo rótulo
df_acao = df_acao.loc[:, ['Adj Close']]

# Renomeando coluna
df_acao.columns = ['PETR4.SA']

# Visualizando resultado
df_acao
```

👉 A partir da base de dados organizada, é fácil exibir um gráfico:

```python
# Criando o gráfico
df_acao.plot()

# Adicionando:
plt.title('Valores de fechamento') # Título
plt.xlabel('Data')                 # Rótulo do eixo horizontal (x)
plt.ylabel('Valor (R$)')           # Rótulo do eixo vertical (y)

# Exibindo o gráfico
plt.show()
```

### Histograma

📊 O histograma é uma representação gráfica que exibe a distribuição de frequência de um conjunto de dados em intervalos ou "blocos".

> 🛠️ *Exemplo prático 2:* Visualizando a distribuição dos valores de uma determinada ação

```python
# Criando o gráfico
df_acao.hist(edgecolor='black',  # Cor das bordas: 'black' = 'preto'
             color = 'purple',   # Cor: 'purple' = 'roxo'
             grid=False          # Linhas de grade desligadas
             )

# Adicionando:
plt.title('Valores de fechamento') # Título
plt.xlabel('Valor (R$)')           # Rótulo do eixo horizontal (x)
plt.ylabel('Contagem')             # Rótulo do eixo vertical (y)

# Exibindo o gráfico
plt.show()
```

> ⛔ O histograma só deve ser utilizado para variáveis com valores numéricos contínuos (ou seja, que podem assumir números decimais como R$ 3,14, por exemplo)

> 📖 A representação acima ilustra a contagem de vezes em que a ação ficou em determinado intervalo. Note que, para essa base de dados neste período de tempo, a ação se estabeleceu mais vezes entre (aproximadamente) R\$ 20,50 e R\$ 21,00 (ou seja, a maior barra do gráfico).

### Gráficos candlestick

🕯️ Também conhecido como gráfico de velas, é uma ferramenta amplamente utilizada na análise técnica em mercados financeiros, como ações, moedas e *commodities*. Sua utilidade inclui:

*  **Representação de preços**: Os *candlesticks* mostram os preços de abertura, fechamento, máxima e mínima de um ativo em um determinado período de tempo, normalmente em forma de velas.

*  **Identificação de padrões**: Os padrões de *candlestick*, como doji, martelo, estrela da manhã, entre outros, ajudam os traders a identificar potenciais reversões de tendência ou continuação de movimentos.

*  **Análise psicológica**: Os *candlesticks* refletem a psicologia do mercado, indicando se os compradores ou vendedores têm controle sobre o preço durante o período analisado.

*  **Tomada de decisões**: Os *traders* usam gráficos de *candlestick* para tomar decisões de compra e venda com base em padrões e tendências identificadas, contribuindo para estratégias de negociação.

*  **Gerenciamento de riscos**: Permitem aos traders estabelecer níveis de suporte e resistência, além de definir ordens para gerenciar o risco de suas operações.

> 🛠️ *Exemplo prático 3:* Visualizando valores de abertura, fechamento, máxima e mínima de uma ação em um determinado período de tempo

> 📖 **TUTORIAL DE INTERPRETAÇÃO DE UM CANDLESTICK**
>
> A interpretação de um *candlestick* envolve a análise dos elementos-chave de cada vela (*candle*) para obter informações sobre a ação do preço em um determinado período de tempo.
>
> **Corpo da Vela**:
> *  O corpo da vela é a parte retangular do *candlestick* e representa a diferença entre os preços de abertura e fechamento durante o período analisado.
>
>   *  Um corpo preenchido ou escuro (geralmente preto ou vermelho) indica que o preço de fechamento é menor do que o preço de abertura.
>
>   *  Um corpo vazio ou claro (geralmente branco ou verde) indica que o preço de fechamento é maior do que o preço de abertura.
>
> **Pavio Superior**:
>
> *  O pavio superior é a linha vertical acima do corpo da vela e representa a máxima alcançada durante o período.
>
> **Pavio Inferior**:
>
> *  O pavio inferior é a linha vertical abaixo do corpo da vela e representa a mínima alcançada durante o período.

### Gráficos de barras

📊 O gráfico de barras é uma representação visual de dados que utiliza barras verticais ou horizontais para mostrar informações quantitativas ou categóricas.

> 🛠️ *Exemplo prático 4:* Comparando a inflação de 2017 a 2021 para os países do BRICS (Fonte: FRED)
*   Inflação - Brasil: https://fred.stlouisfed.org/series/FPCPITOTLZGBRA
*   Inflação - Russia: https://fred.stlouisfed.org/series/FPCPITOTLZGRUS
*   Inflação - Índia: https://fred.stlouisfed.org/series/FPCPITOTLZGIND
*   Inflação - China: https://fred.stlouisfed.org/series/FPCPITOTLZGCHN
*   Inflação - África do Sul: https://fred.stlouisfed.org/series/FPCPITOTLZGZAF

```python
# Carregando pacote
import pandas_datareader.data as web
import datetime

# Definindo código
codigo_fred = ['FPCPITOTLZGBRA', # Inflação - Brasil
               'FPCPITOTLZGRUS', # Inflação - Russia
               'FPCPITOTLZGIND', # Inflação - Índia
               'FPCPITOTLZGCHN', # Inflação - China
               'FPCPITOTLZGZAF'] # Inflação - África do Sul

# Definindo data: datetime.datetime(AAAA, MM, DD)
inicio = datetime.datetime(2017, 1, 1)
fim = datetime.datetime(2021, 1, 1)

# Obtendo dados
df_fred = web.DataReader(codigo_fred, 'fred', inicio, fim)

# Resultado (tabela)
df_fred
```

👉 Vamos organizar o `DataFrame`para que o resultado gráfico seja organizado:

```python
# Modificando as datas só para os anos
df_fred.index = [2017, 2018, 2019, 2020, 2021]

# Modificando os nomes das colunas
df_fred.columns = ['Brasil', 'Russia',
                   'Índia', 'China',
                   'África do Sul']
```

```python
# Criando o gráfico
df_fred.plot.bar(rot=0,             # Angulo da rotação dos rótulos na horizontal
                                    # > 0 = sem rotação
                 edgecolor='black', # Cor das bordas
                 grid = False)      # Remove as linhas de grade

# Adicionando:
plt.title('Inflação (BRICS)')      # Título
plt.xlabel('Ano')                  # Rótulo do eixo horizontal (x)
plt.ylabel('Percentual (%)')       # Rótulo do eixo vertical (y)

# Exibindo o gráfico
plt.show()
```

👉 Podemos personalizar as cores conforme a bandeira nacional de cada país:

```python
# Criando o gráfico
df_fred.plot.bar(rot=0,             # Angulo da rotação dos rótulos na horizontal
                                    # > 0 = sem rotação
                 edgecolor='black', # Cor das bordas
                 grid = False,      # Remove as linhas de grade
                 color = ['gold', 'white',
                          'orange', 'red',
                          'darkgreen'])

# Adicionando:
plt.title('Inflação (BRICS)')      # Título
plt.xlabel('Ano')                  # Rótulo do eixo horizontal (x)
plt.ylabel('Percentual (%)')       # Rótulo do eixo vertical (y)

# Exibindo o gráfico
plt.show()
```

> ⚠️ Gráficos de barras e histogramas são visualizações de dados semelhantes, mas não são a mesma coisa! Eles diferem na forma como representam e são usados para tipos diferentes de dados.
>
> * 👁️‍🗨️ A principal diferença entre gráficos de barras e histogramas está no tipo de dados que eles representam e na forma como são usados:
>   * Gráficos de barras são adequados para dados categóricos ou discretos.
>   * Histogramas são usados para representar a distribuição de dados numéricos ou contínuos.

### Gráficos de linhas (2 ou mais variáveis)

> 🛠️ *Exemplo prático 5:* Visualizando valores de determinadas ações ao longo do tempo
> *  Vamos acompanhar as ações de grandes bancos simultaneamente

```python
# Definindo abreviação (ticker)
bancos_tickers = ["ITUB4.SA",   # Itau
                  "BBDC4.SA",   # Bradesco
                  "SANB11.SA"]  # Santander

# Obtendo dados
df_bancos = yf.download(bancos_tickers,       # Ações
                        start="2023-01-01",   # Data inicial AAAA-MM-DD
                        end="2023-08-01")     # Data final AAAA-MM-DD

# Selecionando as colunas `Adj Close` pela posição
df_bancos = df_bancos.iloc[:, [0, 1, 2]]

# Renomeando colunas
df_bancos.columns = ['ITUB4.SA', 'BBDC4.SA', 'SANB11.SA']

# Criando o gráfico
df_bancos.plot()

# Adicionando:
plt.title('Valores de fechamento') # Título
plt.xlabel('Data')                 # Rótulo do eixo horizontal (x)
plt.ylabel('Valor (R$)')           # Rótulo do eixo vertical (y)

# Exibindo o gráfico
plt.show()
```

💬 Note que as ações aparentam ter um movimento semelhante:
*  ⬆️ Quando uma sobe, as demais sobem,
*  ⬇️ Quando uma desce, as demais descem.

👉 Isso implica que existe alguma *relação* entre elas. Como explorar isso?

### Gráfico de dispersão (2 ou mais variáveis)

🔀 O gráfico de dispersão é uma ferramenta visual usada para representar **pares de dados**, geralmente representando uma variável no eixo horizontal e outra no eixo vertical.

> 🛠️ *Exemplo prático 6:* Visualizando a dispersão conjunta entre pares de ações

```python
# Criando o gráfico
pd.plotting.scatter_matrix(df_bancos)

# Adicionando:
plt.xlabel('Data')                 # Rótulo do eixo horizontal (x)
plt.ylabel('Valor (R$)')           # Rótulo do eixo vertical (y)

# Exibindo o gráfico
plt.show()
```

💬 Note que as dispersões entre os pares são lineares (ou seja, é plausível desenhar uma reta dentro destas nuvens de pontos para descrever a relação entre os pares de ações)

> 📝 Este tipo de relação é essencial para fazermos previsões mais a frente!

> 👁️‍🗨️ Na diagonal *principal*, o histograma é exibido!
