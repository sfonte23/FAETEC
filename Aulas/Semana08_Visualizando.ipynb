## Boas pr√°ticas na cria√ß√£o de gr√°ficos

üé® A cria√ß√£o de gr√°ficos eficazes √© um processo que requer uma compreens√£o profunda tanto dos dados subjacentes quanto dos princ√≠pios de design visual. 2. **Simplicidade e Clareza**: Menos √© mais quando se trata de criar gr√°ficos. Evite o excesso de informa√ß√µes, como r√≥tulos confusos, muitos marcadores ou muitas cores. 3. **Legibilidade dos Dados**: Certifique-se de que os valores exibidos nos gr√°ficos sejam leg√≠veis. 4. **Contextualiza√ß√£o**: Sempre forne√ßa contexto aos dados apresentados. T√≠tulos claros, legendas explicativas e r√≥tulos de eixo bem elaborados ajudam a situar os espectadores no que est√£o vendo, garantindo uma interpreta√ß√£o correta.

5. **Coer√™ncia Visual**: Mantenha uma coer√™ncia visual em todo o gr√°fico e entre diferentes gr√°ficos em uma apresenta√ß√£o. Isso envolve a escolha consistente de paletas de cores, estilos de linhas e formatos de marcadores, de modo que a visualiza√ß√£o seja unificada e f√°cil de seguir.

6. **Destaque de Informa√ß√µes Cruciais**: Se houver informa√ß√µes ou pontos espec√≠ficos que voc√™ deseja destacar, use realces, anota√ß√µes ou elementos visuais para chamar a aten√ß√£o do p√∫blico. Esses destaques podem direcionar o foco para aspectos importantes dos dados.

7. **Minimiza√ß√£o de Distra√ß√µes**: Evite elementos visuais que possam distrair do objetivo principal do gr√°fico. Isso inclui efeitos visuais exagerados, cores desnecess√°rias ou elementos decorativos que n√£o contribuam para a compreens√£o dos dados.

8. **T√≠tulo Explicativo**: Um t√≠tulo claro e conciso ajuda a transmitir a mensagem geral do gr√°fico. Ele deve resumir o que est√° sendo mostrado e destacar os principais insights.

9. **Atualiza√ß√£o e Manuten√ß√£o**: Se os dados subjacentes mudarem com o tempo, certifique-se de atualizar seus gr√°ficos regularmente para refletir as mudan√ßas. Al√©m disso, periodicamente revise e otimize seus gr√°ficos para garantir que permane√ßam eficazes e relevantes.

10. **Teste e Feedback**: Antes de apresentar seus gr√°ficos a um p√∫blico mais amplo, pe√ßa feedback de colegas ou especialistas. Isso pode ajudar a identificar poss√≠veis problemas de interpreta√ß√£o, legibilidade ou formata√ß√£o.

### Gr√°ficos de linhas

üìâ Gr√°ficos de linhas s√£o uma ferramenta valiosa para analisar e comunicar dados, particularmente quando se trata de representar tend√™ncias e mudan√ßas ao longo do tempo.

> üõ†Ô∏è *Exemplo pr√°tico 1:* Visualizando valores de uma determinada a√ß√£o ao longo do tempo

```python
# Carregando pacotes
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Definindo abrevia√ß√£o (ticker)
ticker = ["PETR4.SA"]  # Petrobras SA

# Obtendo dados
df_acao = yf.download(ticker,                # A√ß√£o
                      start="2023-01-01",    # Data inicial AAAA-MM-DD
                      end="2023-04-01")      # Data final AAAA-MM-DD

# Resultado (tabela)
df_acao
```

üëâ Relembrando que s√≥ temos interesse na coluna `Adj Close`
*   Vamos selecionar somente esta coluna (e manter as datas)
*   Vamos renomear para `PETR4.SA`

```python
# Selecionando a coluna `Adj Close` pelo r√≥tulo
df_acao = df_acao.loc[:, ['Adj Close']]

# Renomeando coluna
df_acao.columns = ['PETR4.SA']

# Visualizando resultado
df_acao
```

üëâ A partir da base de dados organizada, √© f√°cil exibir um gr√°fico:

```python
# Criando o gr√°fico
df_acao.plot()

# Adicionando:
plt.title('Valores de fechamento') # T√≠tulo
plt.xlabel('Data')                 # R√≥tulo do eixo horizontal (x)
plt.ylabel('Valor (R$)')           # R√≥tulo do eixo vertical (y)

# Exibindo o gr√°fico
plt.show()
```

üìù **DICA**

> üó£Ô∏è Existem v√°rias formas de personalizar o gr√°fico. Para os gr√°ficos de linhas, algumas op√ß√µes interessantes s√£o adicionar *marcadores (marker)*, *estilo da linha (linestyle)* e a *cor (color)*.

### Histograma

üìä O histograma √© uma representa√ß√£o gr√°fica que exibe a distribui√ß√£o de frequ√™ncia de um conjunto de dados em intervalos ou "blocos".

> üõ†Ô∏è *Exemplo pr√°tico 2:* Visualizando a distribui√ß√£o dos valores de uma determinada a√ß√£o

```python
# Criando o gr√°fico
df_acao.hist(edgecolor='black',  # Cor das bordas: 'black' = 'preto'
             color = 'purple',   # Cor: 'purple' = 'roxo'
             grid=False          # Linhas de grade desligadas
             )

# Adicionando:
plt.title('Valores de fechamento') # T√≠tulo
plt.xlabel('Valor (R$)')           # R√≥tulo do eixo horizontal (x)
plt.ylabel('Contagem')             # R√≥tulo do eixo vertical (y)

# Exibindo o gr√°fico
plt.show()
```

> ‚õî O histograma s√≥ deve ser utilizado para vari√°veis com valores num√©ricos cont√≠nuos (ou seja, que podem assumir n√∫meros decimais como R$ 3,14, por exemplo)

> üìñ A representa√ß√£o acima ilustra a contagem de vezes em que a a√ß√£o ficou em determinado intervalo. Note que, para essa base de dados neste per√≠odo de tempo, a a√ß√£o se estabeleceu mais vezes entre (aproximadamente) R\$ 20,50 e R\$ 21,00 (ou seja, a maior barra do gr√°fico).

*  **Identifica√ß√£o de padr√µes**: Os padr√µes de *candlestick*, como doji, martelo, estrela da manh√£, entre outros, ajudam os traders a identificar potenciais revers√µes de tend√™ncia ou continua√ß√£o de movimentos.

*  **An√°lise psicol√≥gica**: Os *candlesticks* refletem a psicologia do mercado, indicando se os compradores ou vendedores t√™m controle sobre o pre√ßo durante o per√≠odo analisado.

*  **Tomada de decis√µes**: Os *traders* usam gr√°ficos de *candlestick* para tomar decis√µes de compra e venda com base em padr√µes e tend√™ncias identificadas, contribuindo para estrat√©gias de negocia√ß√£o.

*  **Gerenciamento de riscos**: Permitem aos traders estabelecer n√≠veis de suporte e resist√™ncia, al√©m de definir ordens para gerenciar o risco de suas opera√ß√µes.

>

>
>   *  Um corpo vazio ou claro (geralmente branco ou verde) indica que o pre√ßo de fechamento √© maior do que o pre√ßo de abertura.
>
> **Pavio Superior**:
>
> *  O pavio superior √© a linha vertical acima do corpo da vela e representa a m√°xima alcan√ßada durante o per√≠odo.
>
> **Pavio Inferior**:
>
> *  O pavio inferior √© a linha vertical abaixo do corpo da vela e representa a m√≠nima alcan√ßada durante o per√≠odo.

> * üëÅÔ∏è‚Äçüó®Ô∏è A principal diferen√ßa entre gr√°ficos de barras e histogramas est√° no tipo de dados que eles representam e na forma como s√£o usados:

```python
# Definindo abrevia√ß√£o (ticker)
bancos_tickers = ["ITUB4.SA",   # Itau
                  "BBDC4.SA",   # Bradesco
                  "SANB11.SA"]  # Santander

# Obtendo dados
df_bancos = yf.download(bancos_tickers,       # A√ß√µes
                        start="2023-01-01",   # Data inicial AAAA-MM-DD
                        end="2023-08-01")     # Data final AAAA-MM-DD

# Selecionando as colunas `Adj Close` pela posi√ß√£o
df_bancos = df_bancos.iloc[:, [0, 1, 2]]

# Renomeando colunas
df_bancos.columns = ['ITUB4.SA', 'BBDC4.SA', 'SANB11.SA']

# Criando o gr√°fico
df_bancos.plot()

# Adicionando:
plt.title('Valores de fechamento') # T√≠tulo
plt.xlabel('Data')                 # R√≥tulo do eixo horizontal (x)
plt.ylabel('Valor (R$)')           # R√≥tulo do eixo vertical (y)

# Exibindo o gr√°fico
plt.show()
```

### Gr√°fico de dispers√£o (2 ou mais vari√°veis)

üîÄ O gr√°fico de dispers√£o √© uma ferramenta visual usada para representar **pares de dados**, geralmente representando uma vari√°vel no eixo horizontal e outra no eixo vertical.

> üõ†Ô∏è *Exemplo pr√°tico 6:* Visualizando a dispers√£o conjunta entre pares de a√ß√µes

```python
# Criando o gr√°fico
pd.plotting.scatter_matrix(df_bancos)

# Adicionando:
plt.xlabel('Data')                 # R√≥tulo do eixo horizontal (x)
plt.ylabel('Valor (R$)')           # R√≥tulo do eixo vertical (y)

# Exibindo o gr√°fico
plt.show()
```

üí¨ Note que as dispers√µes entre os pares s√£o lineares (ou seja, √© plaus√≠vel desenhar uma reta dentro destas nuvens de pontos para descrever a rela√ß√£o entre os pares de a√ß√µes)

> üìù Este tipo de rela√ß√£o √© essencial para fazermos previs√µes mais a frente!
